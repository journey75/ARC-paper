# EXAALT Maximum Difference Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the maximum difference between original values and their lossy effected counterparts is effected by faults within the different sets of data.

In particular, we will be aiming to answer the following questions:
1. What is the average and std of maximum difference on control and corrupt cases?
2. What percentage of trials violate the error bound by maximum difference?
3. Which bit flips cause the largest maximum difference in the decompressed results?
4. Does an increased maximum difference decrease decompression bandwidth?

## Answers and Conclusions:

In the above analysis we looked at answering the following questions and from our analysis we are able to answer these questions better than before:

1. What is the average and std of maximum difference on control and corrupt cases?
  
The average and std of maximum difference in the control case is relatively small on average with very little variance. However, when we added in faults, the maximum difference skyrockets along with the standard deviation.

2. What percentage of trials violate the error bound by maximum difference?
  
Due to the fact that we are not able to accurately determine the exact error bound for PW and PSNR, we are only able to determine this percentage for ABS. When looking at ABS, we noticed that injecting a fault will violate the error bounds 77-99% of the time for error bounds between 1e4 and 1e2.

3. Which bit flips cause the largest maximum difference in the decompressed results?
  
When looking at each error bounding mode, we were able to see that giant spikes in maximum difference occur towards the earlier bytes in the compressed data with some giant spikes also occuring at smaller error bounding values. It is also worth noting that PW had the highest differences among all.

4. Does an increased maximum difference decrease decompression bandwidth?
  
Overall, decompression bandwidth is effected by maximum difference, but not by as much as one would expect. For instance larger maximum differences did not lead to a lower decompression bandwidth and smaller maximum differences caused larger variations in decompression bandwidths but not on a large scale.
