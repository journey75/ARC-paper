# EXAALT End Status Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the end status changes throughout the trials.

In particular, we will be aiming to answer the following questions:
1. How often does each end status occur?
2. Is there some kind of pattern behind bit flip locations and the statuses they cause?
3. How do these end statuses map to the layout of the compressed data that SZ returns?

## Libraries and Constants

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib
import seaborn as sns
import time

sns.set()
sns.set_style("whitegrid", {'axes.grid' : True})
sns.set_color_codes()
sns.set_context("notebook", font_scale=1.2, rc={"lines.linewidth" : 3.5, "lines.markeredgewidth" : 1., "line.markersize" : 10})
font = {"family" : 'serif'} 
matplotlib.rc('font', **font)
color = ["b", "g", "r", "y", "m"]

A_abs_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_abs_1e3.csv']
A_abs_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_abs_1e3.csv']
B_abs_control_list = ['./notebook_data/hacc_data/sz_hacc_control_abs_1e2.csv', './notebook_data/hacc_data/sz_hacc_control_abs_1e3.csv']
B_abs_corrupt_list = ['./notebook_data/hacc_data/sz_hacc_abs_1e2.csv', './notebook_data/hacc_data/sz_hacc_abs_1e3.csv']

A_pw_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_pw_1e3.csv']
A_pw_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_pw_1e3.csv']
B_pw_control_list = ['./notebook_data/hacc_data/sz_hacc_control_pw_1e2.csv', './notebook_data/hacc_data/sz_hacc_control_pw_1e3.csv']
B_pw_corrupt_list = ['./notebook_data/hacc_data/sz_hacc_pw_1e2.csv', './notebook_data/hacc_data/sz_hacc_pw_1e3.csv']

A_psnr_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_psnr_30.csv', './notebook_data/exaalt_data/sz_exaalt_control_psnr_60.csv']
A_psnr_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_psnr_30.csv', './notebook_data/exaalt_data/sz_exaalt_psnr_60.csv']
B_psnr_control_list = ['./notebook_data/hacc_data/sz_hacc_control_psnr_30.csv', './notebook_data/hacc_data/sz_hacc_control_psnr_60.csv']
B_psnr_corrupt_list = ['./notebook_data/hacc_data/sz_hacc_psnr_30.csv', './notebook_data/hacc_data/sz_hacc_psnr_60.csv']


read_cols=["DataSize", "CompressedSize", "ByteLocation", "FlipLocation", "Status"]
A_data_elements = 2869440
B_data_elements = 280953867

def status_id_generator(row):
    if "Completed" in row['Status']:
        return 0
    elif "SegFault" in row['Status']:
        return 2
    elif "VersionError" in row['Status']:
        return 4
    elif "StepLengthError" in row['Status']:
        return 6
    elif "Timeout" in row['Status']:
        return 8
    elif "CoreDump" in row['Status']:
        return 10
    elif "Unknown" in row['Status']:
        return 12
    else:
        return -1
```

## ABS Analysis

### Read in data

```python
A_control_list = A_abs_control_list
A_corrupt_list = A_abs_corrupt_list

B_control_list = B_abs_control_list
B_corrupt_list = B_abs_corrupt_list

A_error_bounds = ['EXAALT 1e-2', 'EXAALT 1e-3']
B_error_bounds = ['HACC 1e-2', 'HACC 1e-3']

A_control_1 = pd.read_csv(A_control_list[0], usecols=read_cols)
A_control_1.columns = A_control_1.columns.str.strip()
A_control_1['ErrorBound'] = A_error_bounds[0]
A_control_1['StatusID'] = A_control_1.apply(status_id_generator, axis=1)
A_control_2 = pd.read_csv(A_control_list[1], usecols=read_cols)
A_control_2.columns = A_control_2.columns.str.strip()
A_control_2['ErrorBound'] = A_error_bounds[1]
A_control_2['StatusID'] = A_control_2.apply(status_id_generator, axis=1)
A_control = pd.concat([A_control_1, A_control_2], ignore_index=True, sort=False)
A_control['Type'] = 'Control'

A_corrupt_1 = pd.read_csv(A_corrupt_list[0], usecols=read_cols)
A_corrupt_1.columns = A_corrupt_1.columns.str.strip()
A_corrupt_1['ErrorBound'] = A_error_bounds[0]
A_corrupt_1['StatusID'] = A_corrupt_1.apply(status_id_generator, axis=1)
A_corrupt_2 = pd.read_csv(A_corrupt_list[1], usecols=read_cols)
A_corrupt_2.columns = A_corrupt_2.columns.str.strip()
A_corrupt_2['ErrorBound'] = A_error_bounds[1]
A_corrupt_2['StatusID'] = A_corrupt_2.apply(status_id_generator, axis=1)
A_corrupt = pd.concat([A_corrupt_1, A_corrupt_2], ignore_index=True, sort=False)
A_corrupt['Type'] = 'Corrupt'

A_total = pd.concat([A_control, A_corrupt], ignore_index=True, sort=False)

B_control_1 = pd.read_csv(B_control_list[0], usecols=read_cols)
B_control_1.columns = B_control_1.columns.str.strip()
B_control_1['ErrorBound'] = B_error_bounds[0]
B_control_1['StatusID'] = B_control_1.apply(status_id_generator, axis=1)
B_control_2 = pd.read_csv(B_control_list[1], usecols=read_cols)
B_control_2.columns = B_control_2.columns.str.strip()
B_control_2['ErrorBound'] = B_error_bounds[1]
B_control_2['StatusID'] = B_control_2.apply(status_id_generator, axis=1)
B_control = pd.concat([B_control_1, B_control_2], ignore_index=True, sort=False)
B_control['Type'] = 'Control'

B_corrupt_1 = pd.read_csv(B_corrupt_list[0], usecols=read_cols)
B_corrupt_1.columns = B_corrupt_1.columns.str.strip()
B_corrupt_1['ErrorBound'] = B_error_bounds[0]
B_corrupt_1['StatusID'] = B_corrupt_1.apply(status_id_generator, axis=1)
B_corrupt_2 = pd.read_csv(B_corrupt_list[1], usecols=read_cols)
B_corrupt_2.columns = B_corrupt_2.columns.str.strip()
B_corrupt_2['ErrorBound'] = B_error_bounds[1]
B_corrupt_2['StatusID'] = B_corrupt_2.apply(status_id_generator, axis=1)
B_corrupt = pd.concat([B_corrupt_1, B_corrupt_2], ignore_index=True, sort=False)
B_corrupt['Type'] = 'Corrupt'

B_total = pd.concat([B_control, B_corrupt], ignore_index=True, sort=False)

total = pd.concat([A_total, B_total], ignore_index=True, sort=False)
```

### End Status Breakdown

```python
A_percents_1 = A_corrupt_1['Status'].value_counts()/A_corrupt_1.shape[0] * 100
A_status_percents_1 = pd.DataFrame({'Status':A_percents_1.index, 'Percent':A_percents_1.values})
A_status_percents_1['ErrorBound'] = A_error_bounds[0]
A_percents_2 = A_corrupt_2['Status'].value_counts()/A_corrupt_2.shape[0] * 100
A_status_percents_2 = pd.DataFrame({'Status':A_percents_2.index, 'Percent':A_percents_2.values})
A_status_percents_2['ErrorBound'] = A_error_bounds[1]

A_percents = pd.concat([A_status_percents_1, A_status_percents_2], ignore_index=True, sort=False)

B_percents_1 = B_corrupt_1['Status'].value_counts()/B_corrupt_1.shape[0] * 100
B_status_percents_1 = pd.DataFrame({'Status':B_percents_1.index, 'Percent':B_percents_1.values})
B_status_percents_1['ErrorBound'] = B_error_bounds[0]
B_percents_2 = B_corrupt_2['Status'].value_counts()/B_corrupt_2.shape[0] * 100
B_status_percents_2 = pd.DataFrame({'Status':B_percents_2.index, 'Percent':B_percents_2.values})
B_status_percents_2['ErrorBound'] = B_error_bounds[1]

B_percents = pd.concat([B_status_percents_1, B_status_percents_2], ignore_index=True, sort=False)

percents = pd.concat([A_percents, B_percents], ignore_index=True, sort=False)

f, axes = plt.subplots(figsize=(12, 7))
fig = sns.barplot(x="Status",y="Percent", hue="ErrorBound", data=percents, ax=axes)
fig.set_yscale('log')
fig.set_xlabel('Return Status', fontsize=14, weight='bold')
fig.set_ylabel('Percent Of Trials', fontsize=14, weight='bold')
fig.set_title('ABS Return Status Distribution', fontsize=20)
plt.legend(title='ErrorBound',loc='upper right')
f.tight_layout()
plt.setp(fig.get_legend().get_texts(), fontsize='15')
plt.setp(fig.get_legend().get_title(), fontsize='17')
plt.show()
```
