# EXAALT RMSE Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the rmse changes with each of the other metrics.

In particular, we will be aiming to answer the following questions:
1. What is the average and standard deviation of RMSE in control and corrupt cases?
2. How does corruption effect RMSE?
3. Does a higher RMSE correlate with a change in Decompression Bandwidth?
4. Which bit flips cause the highest RMSE?

## ABS Analysis

```python
A_control_list = A_abs_control_list
A_corrupt_list = A_abs_corrupt_list

B_control_list = B_abs_control_list
B_corrupt_list = B_abs_corrupt_list

A_error_bounds = ['1e-2', '1e-3']
B_error_bounds = ['1e-2', '1e-3']

A_control_1 = pd.read_csv(A_control_list[0], usecols=read_cols)
A_control_1.columns = A_control_1.columns.str.strip()
A_control_1['ErrorBound'] = A_error_bounds[0]
A_control_1['DecompressionBandwidth'] = (A_control_1['DataSize']/A_control_1['DecompressionTime'])/1048576
A_control_1['RMSE'] = A_control_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
A_control_2 = pd.read_csv(A_control_list[1], usecols=read_cols)
A_control_2.columns = A_control_2.columns.str.strip()
A_control_2['ErrorBound'] = A_error_bounds[1]
A_control_2['DecompressionBandwidth'] = (A_control_2['DataSize']/A_control_2['DecompressionTime'])/1048576
A_control_2['RMSE'] = A_control_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
A_control = pd.concat([A_control_1, A_control_2], ignore_index=True, sort=False)
A_control['Experiment'] = 'EXAALT Control ABS'

A_corrupt_1 = pd.read_csv(A_corrupt_list[0], usecols=read_cols)
A_corrupt_1.columns = A_corrupt_1.columns.str.strip()
A_corrupt_1['ErrorBound'] = A_error_bounds[0]
A_corrupt_1['DecompressionBandwidth'] = (A_corrupt_1['DataSize']/A_corrupt_1['DecompressionTime'])/1048576
A_corrupt_1['RMSE'] = A_corrupt_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
A_corrupt_2 = pd.read_csv(A_corrupt_list[1], usecols=read_cols)
A_corrupt_2.columns = A_corrupt_2.columns.str.strip()
A_corrupt_2['ErrorBound'] = A_error_bounds[1]
A_corrupt_2['DecompressionBandwidth'] = (A_corrupt_2['DataSize']/A_corrupt_2['DecompressionTime'])/1048576
A_corrupt_2['RMSE'] = A_corrupt_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
A_corrupt = pd.concat([A_corrupt_1, A_corrupt_2], ignore_index=True, sort=False)
A_corrupt['Experiment'] = 'EXAALT Corrupt ABS'

A_total = pd.concat([A_control, A_corrupt], ignore_index=True, sort=False)

B_control_1 = pd.read_csv(B_control_list[0], usecols=read_cols)
B_control_1.columns = B_control_1.columns.str.strip()
B_control_1['ErrorBound'] = B_error_bounds[0]
B_control_1['DecompressionBandwidth'] = (B_control_1['DataSize']/B_control_1['DecompressionTime'])/1048576
B_control_1['RMSE'] = B_control_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
B_control_2 = pd.read_csv(B_control_list[1], usecols=read_cols)
B_control_2.columns = B_control_2.columns.str.strip()
B_control_2['ErrorBound'] = B_error_bounds[1]
B_control_2['DecompressionBandwidth'] = (B_control_2['DataSize']/B_control_2['DecompressionTime'])/1048576
B_control_2['RMSE'] = B_control_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)

B_control = pd.concat([B_control_1, B_control_2], ignore_index=True, sort=False)
B_control['Experiment'] = 'HACC Control ABS'

B_corrupt_1 = pd.read_csv(B_corrupt_list[0], usecols=read_cols)
B_corrupt_1.columns = B_corrupt_1.columns.str.strip()
B_corrupt_1['ErrorBound'] = B_error_bounds[0]
B_corrupt_1['DecompressionBandwidth'] = (B_corrupt_1['DataSize']/B_corrupt_1['DecompressionTime'])/1048576
B_corrupt_1['RMSE'] = B_corrupt_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
B_corrupt_2 = pd.read_csv(B_corrupt_list[1], usecols=read_cols)
B_corrupt_2.columns = B_corrupt_2.columns.str.strip()
B_corrupt_2['ErrorBound'] = B_error_bounds[1]
B_corrupt_2['DecompressionBandwidth'] = (B_corrupt_2['DataSize']/B_corrupt_2['DecompressionTime'])/1048576
B_corrupt_2['RMSE'] = B_corrupt_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
B_corrupt = pd.concat([B_corrupt_1, B_corrupt_2], ignore_index=True, sort=False)
B_corrupt['Experiment'] = 'HACC Corrupt ABS'

B_total = pd.concat([B_control, B_corrupt], ignore_index=True, sort=False)

abs_total = pd.concat([A_total, B_total], ignore_index=True, sort=False)
abs_completed = abs_total[(abs_total["Status"].str.contains('Completed')) | (abs_total["Status"].str.contains('Timeout'))]
abs_corrupt = pd.concat([A_corrupt, B_corrupt], ignore_index=True, sort=False)

exaalt_abs_1e2_corrupt_completed = A_corrupt_1[(A_corrupt_1["Status"].str.contains('Completed'))]
exaalt_abs_1e3_corrupt_completed = A_corrupt_2[(A_corrupt_2["Status"].str.contains('Completed'))]

hacc_abs_1e2_corrupt_completed = B_corrupt_1[(B_corrupt_1["Status"].str.contains('Completed'))]
hacc_abs_1e3_corrupt_completed = B_corrupt_2[(B_corrupt_2["Status"].str.contains('Completed'))]
```

## Answers and Conclusions:

In the above analysis we looked at answering the following questions and from our analysis we are able to answer these questions better than before:

1. What is the average and std number of incorrect elements on control and corrupt cases?
The average and standard deviation of incorrect elements in the control case is 0 as expected. In the corrupt case this value only increases slightly while the standard deviation increases more.

2. How does incorrect count correlate with the other metrics?
In both the decompression bandwidth and maximum difference cases, the effects due to the incorrect elements looked to be on the extreme ends of the spectrum while the effects on RMSE and PSNR seemed evenly distributed.

3. Which bit flips cause the largest percentage of incorrect count?
There looked to be two major regions that caused incorrect elements with a gap between them. In the status analysis section we will be looking at the statuses of each range and attempting to determine what is in this missing range.

4. What percentage of bit flips cause incorrect values?
It is seen that the number of bit flips that cause incorrect values and actually finish increases as the error bound goes up.
