## Step 1: Ingest Data

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import time

sns.set(style="darkgrid")

abs_files = ["./jupyter_notebook_data/sz_abs_1e2_results.csv", "./jupyter_notebook_data/sz_abs_1e3_results.csv", "./jupyter_notebook_data/sz_abs_1e4_results.csv","./jupyter_notebook_data/sz_abs_1e5_results.csv", "./jupyter_notebook_data/sz_abs_1e6_results.csv", "./jupyter_notebook_data/sz_abs_1e7_results.csv"]
pw_files = ["./jupyter_notebook_data/sz_pw_1e2_results.csv", "./jupyter_notebook_data/sz_pw_1e3_results.csv", "./jupyter_notebook_data/sz_pw_1e4_results.csv","./jupyter_notebook_data/sz_pw_1e5_results.csv", "./jupyter_notebook_data/sz_pw_1e6_results.csv", "./jupyter_notebook_data/sz_pw_1e7_results.csv"]
psnr_files = ["./jupyter_notebook_data/sz_psnr_20_results.csv", "./jupyter_notebook_data/sz_psnr_40_results.csv", "./jupyter_notebook_data/sz_psnr_80_results.csv", "./jupyter_notebook_data/sz_psnr_110_results.csv"] # "./jupyter_notebook_data/sz_psnr_60_results.csv" , "./jupyter_notebook_data/sz_psnr_100_results.csv"
test_file_loc = 3
```

```python
# Build Cross Mode DataFrame
mode_data = pd.read_csv(abs_files[test_file_loc], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " MaxDeviation", " RMSE", " PSNR", " Status"])
mode_data.columns = mode_data.columns.str.strip()
mode_data['Mode'] = 'ABS'

temp_data = pd.read_csv(pw_files[test_file_loc], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " MaxDeviation", " RMSE", " PSNR", " Status"])
temp_data.columns = temp_data.columns.str.strip()
temp_data['Mode'] = 'PW'

mode_data = pd.concat([mode_data, temp_data], ignore_index=True)

temp_data = pd.read_csv(psnr_files[test_file_loc], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " MaxDeviation", " RMSE", " PSNR", " Status"])
temp_data.columns = temp_data.columns.str.strip()
temp_data['Mode'] = 'PSNR'

mode_data = pd.concat([mode_data, temp_data], ignore_index=True)
    
# Update Datatypes
mode_data['MaxDifference'] = mode_data['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
mode_data['MaxDeviation'] = mode_data['MaxDeviation'].apply(pd.to_numeric, errors='coerce').fillna(-2)
mode_data['RMSE'] = mode_data['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
mode_data['PSNR'] = mode_data['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)

# Get completed status only
mode_completed = mode_data[(mode_data["Status"].str.contains('Completed'))]
```

```python
print("Mode Data: Length: {} | Columns: {}".format(mode_data.shape[0], mode_data.shape[1]))
#print(mode_data.dtypes)
```

## Step 2: Comparing Different Error Modes

### Incorrect Count vs. Byte Location for each Error Bounding Mode

```python
sns.relplot(x='ByteLocation', y='Incorrect', col="Mode", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=mode_data)
plt.show()
```

### PSNR vs. Byte Location for each Error Bounding Mode

```python
sns.relplot(x='ByteLocation', y='PSNR', col="Mode", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=mode_completed)
plt.show()
```

```python
sns.boxplot(x='Mode', y='PSNR', data=mode_completed).set_title('PSNR Values by Error Bounding Mode')
plt.show()
```

### RMSE vs. Byte Location for each Error Bounding Mode

```python
mode_completed.loc[mode_completed['RMSE'] > 30000, 'RMSE'] = 30000
sns.relplot(x='ByteLocation', y='RMSE', col="Mode", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=mode_completed)
plt.show()
```

### Decompression Time vs. Byte Location for each Error Bounding Mode

```python
sns.relplot(x='ByteLocation', y='DecompressionTime', col="Mode", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=mode_completed)
plt.show()
```

```python
sns.barplot(x='Mode', y='DecompressionTime', data=mode_completed).set_title('Decompression Time by Error Bounding Mode')
plt.show()
```

### Maximum Difference vs. Byte Location for each Error Bounding Mode

```python
sns.relplot(x='ByteLocation', y='MaxDifference', col="Mode", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=mode_completed)
plt.show()
```
