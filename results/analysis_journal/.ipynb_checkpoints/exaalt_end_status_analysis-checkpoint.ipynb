# EXAALT End Status Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the end status changes throughout the trials.

In particular, we will be aiming to answer the following questions:
1. How often does each end status occur?
2. Is there some kind of pattern behind bit flip locations and the statuses they cause?
3. How do these end statuses map to the layout of the compressed data that SZ returns?

## Libraries and Constants

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib
import seaborn as sns
import time

sns.set()
sns.set_style("whitegrid", {'axes.grid' : True})
sns.set_color_codes()
sns.set_context("notebook", font_scale=1.2, rc={"lines.linewidth" : 3.5, "lines.markeredgewidth" : 1., "line.markersize" : 10})
font = {"family" : 'serif'} 
matplotlib.rc('font', **font)
color = ["b", "g", "r", "y", "m"]

abs_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_abs_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_control_abs_1e4.csv']
abs_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_abs_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_abs_1e4.csv']

pw_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_pw_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_control_pw_1e4.csv']
pw_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_pw_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_pw_1e4.csv']

psnr_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_psnr_60.csv', './notebook_data/exaalt_data/sz_exaalt_control_psnr_90.csv', './notebook_data/exaalt_data/sz_exaalt_control_psnr_120.csv']
psnr_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_psnr_60.csv', './notebook_data/exaalt_data/sz_exaalt_psnr_90.csv', './notebook_data/exaalt_data/sz_exaalt_psnr_120.csv']

read_cols=["DataSize", "CompressedSize", "ByteLocation", "FlipLocation", "Status"]
data_elements = 2869440

def status_id_generator(row):
    if "Completed" in row['Status']:
        return 0
    elif "SegFault" in row['Status']:
        return 2
    elif "VersionError" in row['Status']:
        return 4
    elif "StepLengthError" in row['Status']:
        return 6
    elif "Timeout" in row['Status']:
        return 8
    elif "CoreDump" in row['Status']:
        return 10
    elif "Unknown" in row['Status']:
        return 12
    else:
        return -1
```

## ABS Analysis

### Read in data

```python
control_list = abs_control_list
corrupt_list = abs_corrupt_list
error_bounds = ['1e-2', '1e-3', '1e-4']

control_1 = pd.read_csv(control_list[0], usecols=read_cols)
control_1.columns = control_1.columns.str.strip()
control_1['ErrorBound'] = error_bounds[0]
control_1['StatusID'] = control_1.apply(status_id_generator, axis=1)
control_2 = pd.read_csv(control_list[1], usecols=read_cols)
control_2.columns = control_2.columns.str.strip()
control_2['ErrorBound'] = error_bounds[1]
control_2['StatusID'] = control_2.apply(status_id_generator, axis=1)
control_3 = pd.read_csv(control_list[2], usecols=read_cols)
control_3.columns = control_3.columns.str.strip()
control_3['ErrorBound'] = error_bounds[2]
control_3['StatusID'] = control_3.apply(status_id_generator, axis=1)
control = pd.concat([control_1, control_2, control_3], ignore_index=True, sort=False)
control['Type'] = 'Control'

corrupt_1 = pd.read_csv(corrupt_list[0], usecols=read_cols)
corrupt_1.columns = corrupt_1.columns.str.strip()
corrupt_1['ErrorBound'] = error_bounds[0]
corrupt_1['StatusID'] = corrupt_1.apply(status_id_generator, axis=1)
corrupt_2 = pd.read_csv(corrupt_list[1], usecols=read_cols)
corrupt_2.columns = corrupt_2.columns.str.strip()
corrupt_2['ErrorBound'] = error_bounds[1]
corrupt_2['StatusID'] = corrupt_2.apply(status_id_generator, axis=1)
corrupt_3 = pd.read_csv(corrupt_list[2], usecols=read_cols)
corrupt_3.columns = corrupt_3.columns.str.strip()
corrupt_3['ErrorBound'] = error_bounds[2]
corrupt_3['StatusID'] = corrupt_3.apply(status_id_generator, axis=1)
corrupt = pd.concat([corrupt_1, corrupt_2, corrupt_3], ignore_index=True, sort=False)
corrupt['Type'] = 'Corrupt'

total = pd.concat([control, corrupt], ignore_index=True, sort=False)
```

### End Status Breakdown

```python
percents_1 = corrupt_1['Status'].value_counts()/corrupt_1.shape[0] * 100
status_percents_1 = pd.DataFrame({'Status':percents_1.index, 'Percent':percents_1.values})
status_percents_1['ErrorBound'] = error_bounds[0]
percents_2 = corrupt_2['Status'].value_counts()/corrupt_2.shape[0] * 100
status_percents_2 = pd.DataFrame({'Status':percents_2.index, 'Percent':percents_2.values})
status_percents_2['ErrorBound'] = error_bounds[1]
percents_3 = corrupt_3['Status'].value_counts()/corrupt_3.shape[0] * 100
status_percents_3 = pd.DataFrame({'Status':percents_3.index, 'Percent':percents_3.values})
status_percents_3['ErrorBound'] = error_bounds[2]
percents = pd.concat([status_percents_1, status_percents_2, status_percents_3], ignore_index=True, sort=False)

f, axes = plt.subplots(figsize=(12, 7))
fig = sns.barplot(x="Status",y="Percent", hue="ErrorBound", data=percents, ax=axes)
fig.set_yscale('log')
fig.set_xlabel('Return Status', fontsize=14, weight='bold')
fig.set_ylabel('Percent Of Trials', fontsize=14, weight='bold')
fig.set_title('ABS Return Status Distribution', fontsize=20)
f.tight_layout()
plt.setp(fig.get_legend().get_texts(), fontsize='15')
plt.setp(fig.get_legend().get_title(), fontsize='17')
plt.show()
```

```python
start_range = 0
end_range = 80
corrupt_1_bit_ranges = corrupt_1[(corrupt_1["ByteLocation"] >= start_range) & (corrupt_1["ByteLocation"] <= end_range)]
corrupt_1_bit_ranges = corrupt_1_bit_ranges.pivot("FlipLocation", "ByteLocation", "StatusID", )

f, axes = plt.subplots(figsize=(20, 10))
heatmap_colors = ['#FFFFFF', '#800080', '#0000FF', '#40E0D0', '#008000', '#FF0000']
cmap = LinearSegmentedColormap.from_list('Custom', heatmap_colors, len(heatmap_colors))
ax = sns.heatmap(corrupt_1_bit_ranges, vmin=-1, vmax=10, cmap=cmap, ax=axes)
ax.set_title('ABS 1e-2 Status Transitions from {} to {}'.format(start_range, end_range))
colorbar = ax.collections[0].colorbar
colorbar.set_ticks([0, 1.7, 3.6, 5.5, 7.2, 9])
colorbar.set_ticklabels(['Completed', 'SegFault', 'VersionError', 'StepLengthError', 'Timeout', 'CoreDump'])
ax.set_ylabel('Flip Location')
ax.set_xlabel('Byte Location')
ax.invert_yaxis()
_, labels = plt.yticks()
f.tight_layout()
plt.setp(labels, rotation=0)
plt.show()
```

```python
start_range = 0
end_range = 80
corrupt_2_bit_ranges = corrupt_2[(corrupt_2["ByteLocation"] >= start_range) & (corrupt_2["ByteLocation"] <= end_range)]
corrupt_2_bit_ranges = corrupt_2_bit_ranges.pivot("FlipLocation", "ByteLocation", "StatusID", )

f, axes = plt.subplots(figsize=(20, 10))
heatmap_colors = ['#FFFFFF', '#800080', '#0000FF', '#40E0D0', '#008000', '#FF0000']
cmap = LinearSegmentedColormap.from_list('Custom', heatmap_colors, len(heatmap_colors))
ax = sns.heatmap(corrupt_1_bit_ranges, vmin=-1, vmax=10, cmap=cmap, ax=axes)
ax.set_title('ABS 1e-3 Status Transitions from {} to {}'.format(start_range, end_range))
colorbar = ax.collections[0].colorbar
colorbar.set_ticks([0, 1.7, 3.6, 5.5, 7.2, 9])
colorbar.set_ticklabels(['Completed', 'SegFault', 'VersionError', 'StepLengthError', 'Timeout', 'CoreDump'])
ax.set_ylabel('Flip Location')
ax.set_xlabel('Byte Location')
ax.invert_yaxis()
_, labels = plt.yticks()
f.tight_layout()
plt.setp(labels, rotation=0)
plt.show()
```

```python
start_range = 0
end_range = 80
corrupt_3_bit_ranges = corrupt_3[(corrupt_3["ByteLocation"] >= start_range) & (corrupt_3["ByteLocation"] <= end_range)]
corrupt_3_bit_ranges = corrupt_3_bit_ranges.pivot("FlipLocation", "ByteLocation", "StatusID", )

f, axes = plt.subplots(figsize=(20, 10))
heatmap_colors = ['#FFFFFF', '#800080', '#0000FF', '#40E0D0', '#008000', '#FF0000']
cmap = LinearSegmentedColormap.from_list('Custom', heatmap_colors, len(heatmap_colors))
ax = sns.heatmap(corrupt_1_bit_ranges, vmin=-1, vmax=10, cmap=cmap, ax=axes)
ax.set_title('ABS 1e-4 Status Transitions from {} to {}'.format(start_range, end_range))
colorbar = ax.collections[0].colorbar
colorbar.set_ticks([0, 1.7, 3.6, 5.5, 7.2, 9])
colorbar.set_ticklabels(['Completed', 'SegFault', 'VersionError', 'StepLengthError', 'Timeout', 'CoreDump'])
ax.set_ylabel('Flip Location')
ax.set_xlabel('Byte Location')
ax.invert_yaxis()
_, labels = plt.yticks()
f.tight_layout()
plt.setp(labels, rotation=0)
plt.show()
```

## Answers and Conclusions:

In the above analysis we looked at answering the following questions and from our analysis we are able to answer these questions better than before:

1. How often does each end status occur?
  
Above we are able to see how often each return status occurs. Most of the time it does complete, followed by mostly segmentation faults, and then the others.

2. Is there some kind of pattern behind bit flip locations and the statuses they cause?
