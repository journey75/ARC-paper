## Step 1: Ingest Data

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import time

sns.set(style="darkgrid")

abs_files = ["./jupyter_notebook_data/sz_abs_1e2_results.csv", "./jupyter_notebook_data/sz_abs_1e3_results.csv", "./jupyter_notebook_data/sz_abs_1e4_results.csv","./jupyter_notebook_data/sz_abs_1e5_results.csv", "./jupyter_notebook_data/sz_abs_1e6_results.csv", "./jupyter_notebook_data/sz_abs_1e7_results.csv"]
pw_files = ["./jupyter_notebook_data/sz_pw_1e2_results.csv", "./jupyter_notebook_data/sz_pw_1e3_results.csv", "./jupyter_notebook_data/sz_pw_1e4_results.csv","./jupyter_notebook_data/sz_pw_1e5_results.csv", "./jupyter_notebook_data/sz_pw_1e6_results.csv", "./jupyter_notebook_data/sz_pw_1e7_results.csv"]
psnr_files = ["./jupyter_notebook_data/sz_psnr_20_results.csv", "./jupyter_notebook_data/sz_psnr_40_results.csv","./jupyter_notebook_data/sz_psnr_80_results.csv", "./jupyter_notebook_data/sz_psnr_110_results.csv"] # "./jupyter_notebook_data/sz_psnr_60_results.csv", "./jupyter_notebook_data/sz_psnr_100_results.csv"
```

```python
# Build ABS DataFrame
abs_data = pd.read_csv(abs_files[0], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " RMSE", " PSNR", " Status"])
abs_data.columns = abs_data.columns.str.strip()
abs_data['Experiment'] = '1e2'

count = 3
for f in abs_files:
    if '1e2' not in f:
        temp_data = pd.read_csv(f, usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " RMSE", " PSNR", " Status"])
        temp_data.columns = temp_data.columns.str.strip()
        exper_id = '1e{}'.format(count)
        count = count + 1
        temp_data['Experiment'] = exper_id
        abs_data = pd.concat([abs_data, temp_data], ignore_index=True)
    
# Update Datatypes
abs_data['MaxDifference'] = abs_data['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_data['RMSE'] = abs_data['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_data['PSNR'] = abs_data['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)

# Get completed status only
abs_completed = abs_data[(abs_data["Status"].str.contains('Completed'))]
```

```python
# Build PW DataFrame
pw_data = pd.read_csv(pw_files[0], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " RMSE", " PSNR", " Status"])
pw_data.columns = pw_data.columns.str.strip()
pw_data['Experiment'] = '1e2'

count = 3
for f in pw_files:
    if '1e2' not in f:
        temp_data = pd.read_csv(f, usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " RMSE", " PSNR", " Status"])
        temp_data.columns = temp_data.columns.str.strip()
        exper_id = '1e{}'.format(count)
        count = count + 1
        temp_data['Experiment'] = exper_id
        pw_data = pd.concat([pw_data, temp_data], ignore_index=True)
    
# Update Datatypes
pw_data['MaxDifference'] = pw_data['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
pw_data['RMSE'] = pw_data['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
pw_data['PSNR'] = pw_data['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)

# Get completed status only
pw_completed = pw_data[(pw_data["Status"].str.contains('Completed'))]
```

```python
# Build PSNR DataFrame
psnr_list = ["20", "40", "60", "80", "100", "110"]
psnr_data = pd.read_csv(psnr_files[0], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " MaxDifference", " RMSE", " PSNR", " Status"])
psnr_data.columns = psnr_data.columns.str.strip()
psnr_data['Experiment'] = psnr_list[0]

count = 1
for f in psnr_files:
    if '20' not in f:
        temp_data = pd.read_csv(f, usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " MaxDifference", " RMSE", " PSNR", " Status"])
        temp_data.columns = temp_data.columns.str.strip()
        exper_id = psnr_list[count]
        count = count + 1
        temp_data['Experiment'] = exper_id
        psnr_data = pd.concat([psnr_data, temp_data], ignore_index=True)
    
# Update Datatypes
psnr_data['MaxDifference'] = psnr_data['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
psnr_data['RMSE'] = psnr_data['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
psnr_data['PSNR'] = psnr_data['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)

# Get completed status only
psnr_completed = psnr_data[(psnr_data["Status"].str.contains('Completed'))]
```

```python
print("ABS Data: Length: {} | Columns: {}".format(abs_data.shape[0], abs_data.shape[1]))
#print(abs_data.dtypes)
print("PW Data: Length: {} | Columns: {}".format(pw_data.shape[0], pw_data.shape[1]))
#print(pw_data.dtypes)
print("PSNR Data: Length: {} | Columns: {}".format(psnr_data.shape[0], psnr_data.shape[1]))
#print(psnr_data.dtypes)
```

## Step 2: Comparing Different Error Bounding Values for ABS

### Incorrect Count for each Error Bounding Value

```python
sns.relplot(x='ByteLocation', y='Incorrect', col="Experiment", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=abs_data)
plt.show()
```

### PSNR for each Error Bounding Value

```python
sns.relplot(x='ByteLocation', y='PSNR', col="Experiment", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=abs_completed)
plt.show()
```

```python
sns.boxplot(x='Experiment', y='PSNR', data=abs_completed).set_title('ABS: PSNR Values by Error Bounding Value')
plt.show()
```

### RMSE for each Error Bounding Value

```python
abs_completed.loc[abs_completed['RMSE'] > 30000, 'RMSE'] = 30000
sns.relplot(x='ByteLocation', y='RMSE', col="Experiment", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=abs_completed)
plt.show()
```

### Decompression Time for each Error Bounding Value

```python
sns.relplot(x='ByteLocation', y='DecompressionTime', col="Experiment", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=abs_completed)
plt.show()
```

```python
sns.barplot(x='Experiment', y='DecompressionTime', data=abs_completed).set_title('ABS: Decompression Time by Error Bounding Value')
plt.show()
```

### Maximum Difference for each Error Bounding Value

```python
sns.relplot(x='ByteLocation', y='MaxDifference', col="Experiment", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=abs_completed)
plt.show()
```
