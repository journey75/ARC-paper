## Step 1: Libraries and Constants

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import time

sns.set()
sns.set_style("whitegrid", {'axes.grid' : True})
sns.set_color_codes()

sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth" : 3.5, "lines.markeredgewidth" : 1., "line.markersize" : 10})

import matplotlib
color = ["b", "g", "r", "y", "m"]
font = {"family" : 'serif'} 

matplotlib.rc('font', **font)

abs_control_list = ['/Users/dkfulp/Compression_Injection_Experiment/analysis_journal/notebook_data/exaalt_data/sz_exaalt_control_abs_1e2.csv', '/Users/dkfulp/Compression_Injection_Experiment/analysis_journal/notebook_data/exaalt_data/sz_exaalt_control_abs_1e4.csv', '/Users/dkfulp/Compression_Injection_Experiment/analysis_journal/notebook_data/exaalt_data/sz_exaalt_control_abs_1e6.csv']
abs_corrupt_list = ['/Users/dkfulp/Compression_Injection_Experiment/analysis_journal/notebook_data/exaalt_data/sz_exaalt_abs_1e2.csv', '/Users/dkfulp/Compression_Injection_Experiment/analysis_journal/notebook_data/exaalt_data/sz_exaalt_abs_1e4.csv', '/Users/dkfulp/Compression_Injection_Experiment/analysis_journal/notebook_data/exaalt_data/sz_exaalt_abs_1e6.csv']
```

## Step 2: Ingest Data

```python
abs_1e2_control = pd.read_csv(abs_control_list[0], usecols = ["DataSize", "DecompressionTime", "MaxDifference", "RMSE", "PSNR"])
abs_1e2_control.columns = abs_1e2_control.columns.str.strip()
abs_1e2_control['Experiment'] = '1e2'
abs_1e2_control['DecompressionBandwidth'] = (abs_1e2_control['DataSize']/abs_1e2_control['DecompressionTime'])/1048576
abs_1e4_control = pd.read_csv(abs_control_list[1], usecols = ["DataSize", "DecompressionTime", "MaxDifference", "RMSE", "PSNR"])
abs_1e4_control.columns = abs_1e4_control.columns.str.strip()
abs_1e4_control['Experiment'] = '1e4'
abs_1e4_control['DecompressionBandwidth'] = (abs_1e4_control['DataSize']/abs_1e4_control['DecompressionTime'])/1048576
abs_1e6_control = pd.read_csv(abs_control_list[2], usecols = ["DataSize", "DecompressionTime", "MaxDifference", "RMSE", "PSNR"])
abs_1e6_control.columns = abs_1e6_control.columns.str.strip()
abs_1e6_control['Experiment'] = '1e6'
abs_1e6_control['DecompressionBandwidth'] = (abs_1e6_control['DataSize']/abs_1e6_control['DecompressionTime'])/1048576
abs_control = pd.concat([abs_1e2_control, abs_1e4_control, abs_1e6_control], ignore_index=True, sort=False)

abs_1e2_corrupt = pd.read_csv(abs_corrupt_list[0], usecols = ["DataSize", "ByteLocation", "FlipLocation", "DecompressionTime", "Incorrect", "MaxDifference", "RMSE", "PSNR", "Status"])
abs_1e2_corrupt.columns = abs_1e2_corrupt.columns.str.strip()
abs_1e2_corrupt['Experiment'] = '1e2'
abs_1e2_corrupt['DecompressionBandwidth'] = (abs_1e2_corrupt['DataSize']/abs_1e2_corrupt['DecompressionTime'])/1048576
abs_1e4_corrupt = pd.read_csv(abs_corrupt_list[1], usecols = ["DataSize", "ByteLocation", "FlipLocation", "DecompressionTime", "Incorrect", "MaxDifference", "RMSE", "PSNR", "Status"])
abs_1e4_corrupt.columns = abs_1e4_corrupt.columns.str.strip()
abs_1e4_corrupt['Experiment'] = '1e4'
abs_1e4_corrupt['DecompressionBandwidth'] = (abs_1e4_corrupt['DataSize']/abs_1e4_corrupt['DecompressionTime'])/1048576
abs_1e6_corrupt = pd.read_csv(abs_corrupt_list[2], usecols = ["DataSize", "ByteLocation", "FlipLocation", "DecompressionTime", "Incorrect", "MaxDifference", "RMSE", "PSNR", "Status"])
abs_1e6_corrupt.columns = abs_1e6_corrupt.columns.str.strip()
abs_1e6_corrupt['Experiment'] = '1e6'
abs_1e6_corrupt['DecompressionBandwidth'] = (abs_1e6_corrupt['DataSize']/abs_1e6_corrupt['DecompressionTime'])/1048576
abs_corrupt = pd.concat([abs_1e2_corrupt, abs_1e4_corrupt, abs_1e6_corrupt], ignore_index=True, sort=False)
abs_corrupt['MaxDifference'] = abs_corrupt['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_corrupt['RMSE'] = abs_corrupt['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_corrupt['PSNR'] = abs_corrupt['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)

abs_corrupt_completed = abs_corrupt[(abs_corrupt["Status"].str.contains('Completed')) | (abs_corrupt["Status"].str.contains('Timeout'))]
abs_corrupt_completed = abs_corrupt_completed.copy()
```

## Step 3: Comparing Different Error Bounding Values

### Incorrect Count for each Error Bounding Value

```python
rel = sns.relplot(x='ByteLocation', y='Incorrect', col="Experiment", col_wrap=3, data=abs_corrupt)
rel.set(xlim=(-300000, 7000000))
plt.show()
```

```python
sns.boxplot(x='Experiment', y='Incorrect', data=abs_corrupt_completed).set_title('ABS: Incorrect By Error Bounding Value')
plt.show()
```

```python
sns.distplot(abs_corrupt_completed["Incorrect"], bins=7, kde=False).set_title('ABS: Incorrect Distribution')
```

```python
print("ABS 1E2 Average Incorrect: {}".format(abs_1e2_corrupt['Incorrect'].mean()))
print("ABS 1E2 Incorrect Standard Deviation: {}\n".format(abs_1e2_corrupt['Incorrect'].std()))
print("ABS 1E4 Average Incorrect: {}".format(abs_1e4_corrupt['Incorrect'].mean()))
print("ABS 1E4 Incorrect Standard Deviation: {}\n".format(abs_1e4_corrupt['Incorrect'].std()))
print("ABS 1E6 Average Incorrect: {}".format(abs_1e6_corrupt['Incorrect'].mean()))
print("ABS 1E6 Incorrect Standard Deviation: {}\n".format(abs_1e6_corrupt['Incorrect'].std()))
```

### Decompression Time for each Error Bounding Value

```python
cb = sns.boxplot(x='Experiment', y='DecompressionBandwidth', data=abs_control).set_title('ABS: Decompression Bandwidth Control Case')
plt.ylim(None, 500)
plt.show()
```
