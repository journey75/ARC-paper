<h1>Compression Injection Analysis</h1>
  
<h2>Introduction</h2>
<p>As HPC system continue to grow in size, the data being used with these systems is continuing to grow in size as well. However, as this data grows a bottleneck occurs when trying to move this vast amount of data around the network. To overcome this bottleneck, compression has been used. Majority of compression utilizes lossless compressors which ensure the full data's integrity but sacrifices a possibly better compression ratio.  Other works have shown that lossy compression can further push the compression ratio at the cost of some data percision.  
  
When a lossy compressor is used to compress data, similar chunks of data are grouped and only a single copy of the data is kept. This process can push the compression ratio further past that of lossless compressors, but does reveal some vulerabilities as well. In this study, we aim to gauge the impacts of faults on lossy compressed data and determine where future protections could be added to further protect the sensitive data.</p>
    
<p>In our experiment, we will be looking at how faults effect different aspects of a lossy compressor. Specifically we will be looking at how faults effect the decompression time, number of incorrect elements, maximum deviation, maximum difference, RMSE, PSNR, and return status.  
  
To test these different aspects we will be compressing and decompressing the sz test float data which is comprised of 8192 single percision floats, the x positional information from the EXAALT molecular dynamics simulation which is comprised of 2869440 single percision floats, and the x positional information from the HACC particle simulation's data found on SDRBench which is comprised of 280953867 single percision floats.  
  
To perform our tests, we run an extensive sweep of 10% of the byte positions within the compressed data. At each of these positions, we will inject exhaustively into each bit found here and record the results. This test will be performed repeatedly while varying the error bounding information for abs, pw relatvie, and psnr within sz.  
  
Upon completion of these tests, a visualization of these can be seen below:</p>

<h2>Experiment Analysis</h2>

<p>To begin our analysis we will first import the nessecary libraries to visualize our tests. In this analysis we will be using seaborn, pandas, and matplotlib.</p>

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import time

sns.set(style="darkgrid")
```

<h3>Ingest Data</h3>

```python
abs_files = ["./jupyter_notebook_data/sz_abs_1e2_results.csv", "./jupyter_notebook_data/sz_abs_1e3_results.csv", "./jupyter_notebook_data/sz_abs_1e4_results.csv","./jupyter_notebook_data/sz_abs_1e5_results.csv", "./jupyter_notebook_data/sz_abs_1e6_results.csv", "./jupyter_notebook_data/sz_abs_1e7_results.csv"]

# Build DataFrame
abs_data = pd.read_csv(abs_files[0], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " MaxDeviation", " RMSE", " PSNR", " Status"])
abs_data.columns = abs_data.columns.str.strip()
abs_data['Experiment'] = '1e2'

count = 3
for f in abs_files:
    if '1e2' not in f:
        temp_data = pd.read_csv(f, usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " MaxDeviation", " RMSE", " PSNR", " Status"])
        temp_data.columns = temp_data.columns.str.strip()
        exper_id = '1e{}'.format(count)
        count = count + 1
        temp_data['Experiment'] = exper_id
        abs_data = pd.concat([abs_data, temp_data], ignore_index=True)
    
# Update Datatypes
abs_data['MaxDifference'] = abs_data['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_data['MaxDeviation'] = abs_data['MaxDeviation'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_data['RMSE'] = abs_data['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_data['PSNR'] = abs_data['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)

print("Abs Data: Length: {} | Columns: {}".format(abs_data.shape[0], abs_data.shape[1]))
#print(abs_data.dtypes)
```

<h3>Comparing Different Error Bounding Approaches</h3>

<h4>1: Incorrect Count vs Byte Location for each data set</h4>
  
<p>To initially visualize the data, we can start by displaying the Byte and Bit Flip Locations for each different 
    experiment.</p>

```python
sns.relplot(x='ByteLocation', y='Incorrect', col="Experiment", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=abs_data)
plt.show()
```

When looking at all of the different experiments we can notice a few things initially. First we can see that the number of byte flip locations increases with more percise error bounding information. This is expected since at more percise error bounds the data cannot be collapsed as much as it would be with less percise error bounding information. This can be seen when comparing the slopes of 1e2 and 1e7. Another interesting observation is with higher error bounding information, we can see a linear pattern at the end of each slope. These values are assumed to be values that could not be compressed. This makes sense since injecting faults into them do not cause large numbers of incorrect values.

<h4>2: Comparing Decompression Time Across Error Bounding Information</h4>
  
<p>Next we can compare how the decompression time varies among each of the different error bounding sets</p>

<p>First we will need to filter and grab only completed end status runs since these are the only ones whose end time  will make sense to visualize</p>

We can also see that as the error bounding becomes more strict, the decompression time rises as well. <h3>3: Incorrect Count vs Byte Location for 1e6</h3>
  
<p>Taking one of the datasets, we can visualize how the Byte Location and Flip Location effect the number of incorrect and ending status of compression.</p>

```python
single_abs_data = abs_data[abs_data['Experiment'].str.contains("1e6")]

sns.relplot(x='ByteLocation', y='Incorrect', col="Status", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=single_abs_data)
plt.show()
```

Upon looking at a single experiment we can see a few interesting trends. First we notice that the previously seen slopes have consisted primarily of completed values. This is expected since this is nessecary to gather the number of incorrect value. Next we can see that VersionError, StepLengthError, Timeout, and CoreDump errors are very rare. One may wonder how these values change with different data formats or lengths. This is something we will discuss later. The last interesting trend we can notice is the relation between SegFaults and the curve. From a first glance it looks as though the SegFaults occur periodically and stop at the same byte location as the end of the completed slope. From here we can visualize this and see if this holds true.

<h3>4: Incorrect Count vs Byte Location for 1e6's Completed and SegFault Ststuses</h3>
  
<p>Taking one of the datasets, we can visualize how the Byte Location and Flip Location effect the number of incorrect and ending status of compression.</p>

```python
completed_v_segfault = single_abs_data[(single_abs_data["Status"].str.contains('Completed')) | (single_abs_data["Status"].str.contains('SegFault'))]

sns.relplot(x='ByteLocation', y='Incorrect', hue="Status", palette=['#008000', '#FF0000'], data=completed_v_segfault)
plt.show()
```

By overlaying these two statuses we can see that the SegFaults seem to correspond with the end of the curve.  This curve is possibly the range of reused values in the compression process.  Once these values are not being reused (i.e the end of the curve) the segmentation faults seem to die off. Now that this relation is known we can look into other metrics such as RMSE and PSNR.
