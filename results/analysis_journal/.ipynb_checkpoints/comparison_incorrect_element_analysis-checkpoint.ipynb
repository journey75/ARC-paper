# EXAALT Incorrect Element Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the incorrect percentage changes with each of the other metrics.

In particular, we will be aiming to answer the following questions:
1. What is the average and std number of incorrect elements on control and corrupt cases?
2. How does incorrect count correlate with the other metrics?
3. Which bit flips cause the largest percentage of incorrect count?
4. What percentage of bit flips cause incorrect values?
4. Does an increase in incorrect elements decrease decompression bandwidth?
