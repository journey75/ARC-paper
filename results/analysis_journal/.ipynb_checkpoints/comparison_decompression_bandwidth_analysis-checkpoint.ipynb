# EXAALT Decompression Bandwidth Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the decompression bandwidth is effected by faults within the different sets of data.

In particular, we will be aiming to answer the following questions:
1. What is the average and std of decompression bandwidth on control and corrupt cases?
2. How does this average compare to the memory system bandwidth as seen by STREAM (Using same size as our data)
3. How does corruption effect decompression bandwidth?
4. Which bit flips cause the largest decrease in decompression bandwidth?

## ABS Analysis

### Read in data

```python
abs_1e2_control = pd.read_csv(abs_control_list[0], usecols=read_cols)
abs_1e2_control.columns = abs_1e2_control.columns.str.strip()
abs_1e2_control['ErrorBound'] = '1e-2'
abs_1e2_control['DecompressionBandwidth'] = (abs_1e2_control['DataSize']/abs_1e2_control['DecompressionTime'])/1048576
abs_1e3_control = pd.read_csv(abs_control_list[1], usecols=read_cols)
abs_1e3_control.columns = abs_1e3_control.columns.str.strip()
abs_1e3_control['ErrorBound'] = '1e-3'
abs_1e3_control['DecompressionBandwidth'] = (abs_1e3_control['DataSize']/abs_1e3_control['DecompressionTime'])/1048576
abs_1e4_control = pd.read_csv(abs_control_list[2], usecols=read_cols)
abs_1e4_control.columns = abs_1e4_control.columns.str.strip()
abs_1e4_control['ErrorBound'] = '1e-4'
abs_1e4_control['DecompressionBandwidth'] = (abs_1e4_control['DataSize']/abs_1e4_control['DecompressionTime'])/1048576
abs_control = pd.concat([abs_1e2_control, abs_1e3_control, abs_1e4_control], ignore_index=True, sort=False)
abs_control['Type'] = 'Control'
abs_control['Experiment'] = 'Control ABS'

abs_1e2_corrupt = pd.read_csv(abs_corrupt_list[0], usecols=read_cols)
abs_1e2_corrupt.columns = abs_1e2_corrupt.columns.str.strip()
abs_1e2_corrupt['ErrorBound'] = '1e-2'
abs_1e2_corrupt['DecompressionBandwidth'] = (abs_1e2_corrupt['DataSize']/abs_1e2_corrupt['DecompressionTime'])/1048576
abs_1e2_corrupt['IncorrectPercentage'] = (abs_1e2_corrupt['Incorrect']/2869440)*100
abs_1e3_corrupt = pd.read_csv(abs_corrupt_list[1], usecols=read_cols)
abs_1e3_corrupt.columns = abs_1e3_corrupt.columns.str.strip()
abs_1e3_corrupt['ErrorBound'] = '1e-3'
abs_1e3_corrupt['DecompressionBandwidth'] = (abs_1e3_corrupt['DataSize']/abs_1e3_corrupt['DecompressionTime'])/1048576
abs_1e3_corrupt['IncorrectPercentage'] = (abs_1e3_corrupt['Incorrect']/2869440)*100
abs_1e4_corrupt = pd.read_csv(abs_corrupt_list[2], usecols=read_cols)
abs_1e4_corrupt.columns = abs_1e4_corrupt.columns.str.strip()
abs_1e4_corrupt['ErrorBound'] = '1e-4'
abs_1e4_corrupt['DecompressionBandwidth'] = (abs_1e4_corrupt['DataSize']/abs_1e4_corrupt['DecompressionTime'])/1048576
abs_1e4_corrupt['IncorrectPercentage'] = (abs_1e4_corrupt['Incorrect']/2869440)*100
abs_corrupt = pd.concat([abs_1e2_corrupt, abs_1e3_corrupt, abs_1e4_corrupt], ignore_index=True, sort=False)
abs_corrupt['Type'] = 'Corrupt'
abs_corrupt['Experiment'] = 'Corrupt ABS'

abs_total = pd.concat([abs_control, abs_corrupt], ignore_index=True, sort=False)
abs_completed = abs_total[(abs_total["Status"].str.contains('Completed')) | (abs_total["Status"].str.contains('Timeout'))]
```

### Control vs. Corrupt Decompression Bandwidth 

```python
sns.barplot(x='ErrorBound', y='DecompressionBandwidth', hue="Type", data=abs_total).set_title('ABS Decompression Bandwidth by Error Bound', fontsize=15, weight='bold')
plt.xlabel("Error Bound",fontsize=14, weight='bold')
plt.ylabel("Decompression\nBandwidth (MB/s)",fontsize=14,weight='bold')
plt.tight_layout()
plt.show()
```

### Control Decompression Bandwidth Breakdown

```python
print("ABS 1E2 Average Decompression Bandwidth: {} MB/s".format(abs_1e2_control['DecompressionBandwidth'].mean()))
print("ABS 1E2 Decompression Bandwidth Standard Deviation: {} MB/s\n".format(abs_1e2_control['DecompressionBandwidth'].std()))
print("ABS 1E3 Average Decompression Bandwidth: {} MB/s".format(abs_1e3_control['DecompressionBandwidth'].mean()))
print("ABS 1E3 Decompression Bandwidth Standard Deviation: {} MB/s\n".format(abs_1e3_control['DecompressionBandwidth'].std()))
print("ABS 1E4 Average Decompression Bandwidth: {} MB/s".format(abs_1e4_control['DecompressionBandwidth'].mean()))
print("ABS 1E4 Decompression Bandwidth Standard Deviation: {} MB/s\n".format(abs_1e4_control['DecompressionBandwidth'].std()))
```

### Corrupt Decompression Bandwidth Breakdown

```python
abs_1e2_corrupt_completed = abs_1e2_corrupt[(abs_1e2_corrupt["Status"].str.contains('Completed')) | (abs_1e2_corrupt["Status"].str.contains('Timeout'))]
abs_1e3_corrupt_completed = abs_1e3_corrupt[(abs_1e3_corrupt["Status"].str.contains('Completed')) | (abs_1e3_corrupt["Status"].str.contains('Timeout'))]
abs_1e4_corrupt_completed = abs_1e4_corrupt[(abs_1e4_corrupt["Status"].str.contains('Completed')) | (abs_1e4_corrupt["Status"].str.contains('Timeout'))]

print("ABS 1E2 Average Decompression Bandwidth: {} MB/s".format(abs_1e2_corrupt_completed['DecompressionBandwidth'].mean()))
print("ABS 1E2 Decompression Bandwidth Standard Deviation: {} MB/s\n".format(abs_1e2_corrupt_completed['DecompressionBandwidth'].std()))
print("ABS 1E3 Average Decompression Bandwidth: {} MB/s".format(abs_1e3_corrupt_completed['DecompressionBandwidth'].mean()))
print("ABS 1E3 Decompression Bandwidth Standard Deviation: {} MB/s\n".format(abs_1e3_corrupt_completed['DecompressionBandwidth'].std()))
print("ABS 1E4 Average Decompression Bandwidth: {} MB/s".format(abs_1e4_corrupt_completed['DecompressionBandwidth'].mean()))
print("ABS 1E4 Decompression Bandwidth Standard Deviation: {} MB/s\n".format(abs_1e4_corrupt_completed['DecompressionBandwidth'].std()))
```

When looking at these averages we notice that the decompression time overall does not seem to be affected by corruption. However, by looking at the standard deviation on each of them we notice that there is a higher variance overall which could indicate some outliers. We will make a boxplot to see if we can find these outliers.

### Corrupt Decompression Bandwidth Distribution

```python
f, axes = plt.subplots(2, 1, figsize=(20, 7))
sns.boxplot(x='ErrorBound', y='DecompressionBandwidth', hue="Type", data=abs_completed, ax=axes[0])
sns.boxplot(x='ErrorBound', y='DecompressionBandwidth', hue="Type", data=abs_completed, ax=axes[1])
plt.ylim(10, 100)
plt.yscale('log')
#plt.ylim(74, 76)
plt.show()
```

As expected, there are some outliers that have extremely high bandwidth values as well as some with much lower bandwidth values as well. From here we can look at how many incorrect elements each of these extremely high bandwidth runs ended up with.

### Corrupt Decompression Bandwidth Outlier Analysis

```python
abs_1e2_corrupt_upper_outliers = abs_1e2_corrupt_completed[abs_1e2_corrupt_completed['DecompressionBandwidth'] > abs_1e2_corrupt_completed.DecompressionBandwidth.quantile(0.75)]
abs_1e2_corrupt_lower_outliers = abs_1e2_corrupt_completed[abs_1e2_corrupt_completed['DecompressionBandwidth'] < abs_1e2_corrupt_completed.DecompressionBandwidth.quantile(0.25)]
abs_1e3_corrupt_upper_outliers = abs_1e3_corrupt_completed[abs_1e3_corrupt_completed['DecompressionBandwidth'] > abs_1e3_corrupt_completed.DecompressionBandwidth.quantile(0.75)]
abs_1e3_corrupt_lower_outliers = abs_1e3_corrupt_completed[abs_1e3_corrupt_completed['DecompressionBandwidth'] < abs_1e3_corrupt_completed.DecompressionBandwidth.quantile(0.25)]
abs_1e4_corrupt_upper_outliers = abs_1e4_corrupt_completed[abs_1e4_corrupt_completed['DecompressionBandwidth'] > abs_1e4_corrupt_completed.DecompressionBandwidth.quantile(0.75)]
abs_1e4_corrupt_lower_outliers = abs_1e4_corrupt_completed[abs_1e4_corrupt_completed['DecompressionBandwidth'] < abs_1e4_corrupt_completed.DecompressionBandwidth.quantile(0.25)]
```

```python
sns.scatterplot(x='DecompressionBandwidth', y='IncorrectPercentage', data=abs_1e2_corrupt_upper_outliers).set_title('ABS 1e2: Incorrect by Decompression Bandwidth > Q3')
plt.show()
sns.scatterplot(x='DecompressionBandwidth', y='IncorrectPercentage', data=abs_1e3_corrupt_upper_outliers).set_title('ABS 1e3: Incorrect by Decompression Bandwidth > Q3')
plt.show()
sns.scatterplot(x='DecompressionBandwidth', y='IncorrectPercentage', data=abs_1e4_corrupt_upper_outliers).set_title('ABS 1e4: Incorrect by Decompression Bandwidth > Q3')
plt.show()
```

When looking at the upper quartile, we notice a few things. First we can see that there are only a few outliers in total that are distinct from the others that are above the third quartile. We can also see that within this group, the maximum number of incorrects are found. This leads us to believe that if a decompression run has an absurdly high decompression bandwidth, it may be safe to assume that its values are incorrect.  
  
Let's now look at the lower quartile.

```python
sns.scatterplot(x='DecompressionBandwidth', y='IncorrectPercentage', data=abs_1e2_corrupt_lower_outliers).set_title('ABS 1e2: Incorrect by Decompression Bandwidth < Q1')
plt.show()
sns.scatterplot(x='DecompressionBandwidth', y='IncorrectPercentage', data=abs_1e3_corrupt_lower_outliers).set_title('ABS 1e3: Incorrect by Decompression Bandwidth < Q1')
plt.show()
sns.scatterplot(x='DecompressionBandwidth', y='IncorrectPercentage', data=abs_1e4_corrupt_lower_outliers).set_title('ABS 1e4: Incorrect by Decompression Bandwidth < Q1')
plt.show()
```

Unlike in the upper quartile cases, the number of incorrect can vary drastically when looking at items lower than the first quartile.

## PSNR Analysis

```python
psnr_60_control = pd.read_csv(psnr_control_list[0], usecols=read_cols)
psnr_60_control.columns = psnr_60_control.columns.str.strip()
psnr_60_control['ErrorBound'] = '060'
psnr_60_control['DecompressionBandwidth'] = (psnr_60_control['DataSize']/psnr_60_control['DecompressionTime'])/1048576
psnr_90_control = pd.read_csv(psnr_control_list[1], usecols=read_cols)
psnr_90_control.columns = psnr_90_control.columns.str.strip()
psnr_90_control['ErrorBound'] = '090'
psnr_90_control['DecompressionBandwidth'] = (psnr_90_control['DataSize']/psnr_90_control['DecompressionTime'])/1048576
psnr_120_control = pd.read_csv(psnr_control_list[2], usecols=read_cols)
psnr_120_control.columns = psnr_120_control.columns.str.strip()
psnr_120_control['ErrorBound'] = '120'
psnr_120_control['DecompressionBandwidth'] = (psnr_120_control['DataSize']/psnr_120_control['DecompressionTime'])/1048576
psnr_control = pd.concat([psnr_60_control, psnr_90_control, psnr_120_control], ignore_index=True, sort=False)
psnr_control['Type'] = 'Control'
psnr_control['Experiment'] = 'Control PSNR'

psnr_60_corrupt = pd.read_csv(psnr_corrupt_list[0], usecols=read_cols)
psnr_60_corrupt.columns = psnr_60_corrupt.columns.str.strip()
psnr_60_corrupt['ErrorBound'] = '060'
psnr_60_corrupt['DecompressionBandwidth'] = (psnr_60_corrupt['DataSize']/psnr_60_corrupt['DecompressionTime'])/1048576
psnr_90_corrupt = pd.read_csv(psnr_corrupt_list[1], usecols=read_cols)
psnr_90_corrupt.columns = psnr_90_corrupt.columns.str.strip()
psnr_90_corrupt['ErrorBound'] = '090'
psnr_90_corrupt['DecompressionBandwidth'] = (psnr_90_corrupt['DataSize']/psnr_90_corrupt['DecompressionTime'])/1048576
psnr_120_corrupt = pd.read_csv(psnr_corrupt_list[2], usecols=read_cols)
psnr_120_corrupt.columns = psnr_120_corrupt.columns.str.strip()
psnr_120_corrupt['ErrorBound'] = '120'
psnr_120_corrupt['DecompressionBandwidth'] = (psnr_120_corrupt['DataSize']/psnr_120_corrupt['DecompressionTime'])/1048576
psnr_corrupt = pd.concat([psnr_60_corrupt, psnr_90_corrupt, psnr_120_corrupt], ignore_index=True, sort=False)
psnr_corrupt['Type'] = 'Corrupt'
psnr_corrupt['Experiment'] = 'Corrupt PSNR'

psnr_corrupt_completed = psnr_corrupt[(psnr_corrupt["Status"].str.contains('Completed')) | (psnr_corrupt["Status"].str.contains('Timeout'))]
psnr_corrupt_completed = psnr_corrupt_completed.copy()

psnr_total = pd.concat([psnr_control, psnr_corrupt], ignore_index=True, sort=False)
```

```python
sns.barplot(x='ErrorBound', y='DecompressionBandwidth', hue="Type", data=psnr_total).set_title('Decompression Bandwidth by Error Bound')
plt.show()
```
