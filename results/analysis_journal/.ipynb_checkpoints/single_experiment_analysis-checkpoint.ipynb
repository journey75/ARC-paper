## Step 1: Ingest Data

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import time

sns.set(style="darkgrid")

abs_files = ["./jupyter_notebook_data/sz_abs_1e2_results.csv", "./jupyter_notebook_data/sz_abs_1e3_results.csv", "./jupyter_notebook_data/sz_abs_1e4_results.csv","./jupyter_notebook_data/sz_abs_1e5_results.csv", "./jupyter_notebook_data/sz_abs_1e6_results.csv", "./jupyter_notebook_data/sz_abs_1e7_results.csv"]
pw_files = ["./jupyter_notebook_data/sz_pw_1e2_results.csv", "./jupyter_notebook_data/sz_pw_1e3_results.csv", "./jupyter_notebook_data/sz_pw_1e4_results.csv","./jupyter_notebook_data/sz_pw_1e5_results.csv", "./jupyter_notebook_data/sz_pw_1e6_results.csv", "./jupyter_notebook_data/sz_pw_1e7_results.csv"]
psnr_files = ["./jupyter_notebook_data/sz_psnr_20_results.csv", "./jupyter_notebook_data/sz_psnr_40_results.csv", "./jupyter_notebook_data/sz_psnr_80_results.csv", "./jupyter_notebook_data/sz_psnr_110_results.csv"] # "./jupyter_notebook_data/sz_psnr_60_results.csv" , "./jupyter_notebook_data/sz_psnr_100_results.csv"
test_file_loc = 3
```

```python
# Build Cross Mode DataFrame
data = pd.read_csv(abs_files[test_file_loc], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " MaxDeviation", " RMSE", " PSNR", " Status"])
#mode_data = pd.read_csv(pw_files[test_file_loc], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " MaxDeviation", " RMSE", " PSNR", " Status"])
#mode_data = pd.read_csv(psnr_files[test_file_loc], usecols = [" ByteLocation", " FlipLocation", " DecompressionTime", " Incorrect", " MaxDifference", " MaxDeviation", " RMSE", " PSNR", " Status"])
data.columns = data.columns.str.strip()

# Update Datatypes
data['MaxDifference'] = data['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
data['MaxDeviation'] = data['MaxDeviation'].apply(pd.to_numeric, errors='coerce').fillna(-2)
data['RMSE'] = data['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
data['PSNR'] = data['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)

# Get completed status only
completed_only = data[(data["Status"].str.contains('Completed'))]
```

```python
print("Data: Length: {} | Columns: {}".format(data.shape[0], data.shape[1]))
#print(data.dtypes)
```

## Step 2: Visualizing Data

### Incorrect Count vs. Byte Location

```python
sns.relplot(x='ByteLocation', y='Incorrect', hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=data)
plt.show()
```

### Incorrect Count vs. Byte Location by End Status

```python
sns.relplot(x='ByteLocation', y='Incorrect', col="Status", col_wrap=3, hue="FlipLocation", palette=['#CC99A2', '#800080', '#0000FF', '#40E0D0', '#008000', '#E59400', '#FF0000', '#000000'], data=data)
plt.show()
```

### Incorrect Count vs. Byte Location for Completed and Segmentation Faults

```python
completed_v_segfault = data[(data["Status"].str.contains('Completed')) | (data["Status"].str.contains('SegFault'))]
sns.relplot(x='ByteLocation', y='Incorrect', hue="Status", palette=['#008000', '#FF0000'], data=completed_v_segfault)
plt.show()
```

### Status vs. Byte Location for All Other Statuses

```python
other_statuses = data[(data["Status"].str.contains('VersionError')) | (data["Status"].str.contains('StepLengthError')) | (data["Status"].str.contains('Timeout')) | (data["Status"].str.contains('CoreDump'))]
sns.relplot(x='ByteLocation', y='Status', data=other_statuses)
plt.show()
```

### Status Transitions by Byte and Flip Location

```python
def status_id_gen(row):
    if "Completed" in row['Status']:
        return 0
    elif "SegFault" in row['Status']:
        return 2
    elif "VersionError" in row['Status']:
        return 4
    elif "StepLengthError" in row['Status']:
        return 6
    elif "Timeout" in row['Status']:
        return 8
    elif "CoreDump" in row['Status']:
        return 10
    else:
        return -1

start_range = 0
end_range = 400
transition_data = data.copy()
transition_data = transition_data.loc[(transition_data['ByteLocation'] >= start_range) & (transition_data['ByteLocation'] <= end_range)]
transition_data['StatusID'] = transition_data.apply(status_id_gen, axis=1)
transition_data = transition_data.pivot("FlipLocation", "ByteLocation", "StatusID")
heatmap_colors = ['#FFFFFF', '#800080', '#0000FF', '#40E0D0', '#008000', '#FF0000']
cmap = LinearSegmentedColormap.from_list('Custom', heatmap_colors, len(heatmap_colors))
ax = sns.heatmap(transition_data, vmin=-1, vmax=10, cmap=cmap)
ax.set_title('Status Transitions from {} to {}'.format(start_range, end_range))
colorbar = ax.collections[0].colorbar
colorbar.set_ticks([0, 1.7, 3.6, 5.5, 7.2, 9])
colorbar.set_ticklabels(['Completed', 'SegFault', 'VersionError', 'StepLengthError', 'Timeout', 'CoreDump'])
ax.set_ylabel('Flip Location')
ax.set_xlabel('Byte Location')
_, labels = plt.yticks()
plt.setp(labels, rotation=0)
plt.show()
```
