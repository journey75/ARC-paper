# EXAALT Maximum Difference Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the maximum difference between original values and their lossy effected counterparts is effected by faults within the different sets of data.

In particular, we will be aiming to answer the following questions:
1. What is the average and std of maximum difference on control and corrupt cases?
2. What percentage of trials violate the error bound by maximum difference?
3. Which bit flips cause the largest maximum difference in the decompressed results?
4. Does an increased maximum difference decrease decompression bandwidth?

## Libraries and Constants

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns
import time

sns.set()
sns.set_style("whitegrid", {'axes.grid' : True})
sns.set_color_codes()
sns.set_context("notebook", font_scale=1.1, rc={"lines.linewidth" : 3.5, "lines.markeredgewidth" : 1., "line.markersize" : 10})
font = {"family" : 'serif'} 
matplotlib.rc('font', **font)
color = ["b", "g", "r", "y", "m"]

abs_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_abs_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_control_abs_1e4.csv']
abs_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_abs_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_abs_1e4.csv']

pw_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_pw_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_control_pw_1e4.csv']
pw_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_pw_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_pw_1e4.csv']

psnr_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_psnr_60.csv', './notebook_data/exaalt_data/sz_exaalt_control_psnr_90.csv', './notebook_data/exaalt_data/sz_exaalt_control_psnr_120.csv']
psnr_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_psnr_60.csv', './notebook_data/exaalt_data/sz_exaalt_psnr_90.csv', './notebook_data/exaalt_data/sz_exaalt_psnr_120.csv']

read_cols=["DataSize", "ByteLocation", "FlipLocation", "DecompressionTime", "Incorrect", "MaxDifference", "Status"]
```

## ABS Analysis

### Read in data

```python
abs_1e2_control = pd.read_csv(abs_control_list[0], usecols=read_cols)
abs_1e2_control.columns = abs_1e2_control.columns.str.strip()
abs_1e2_control['ErrorBound'] = '1e-2'
abs_1e2_control['DecompressionBandwidth'] = (abs_1e2_control['DataSize']/abs_1e2_control['DecompressionTime'])/1048576
abs_1e2_control['IncorrectPercentage'] = (abs_1e2_control['Incorrect']/2869440)*100
abs_1e3_control = pd.read_csv(abs_control_list[1], usecols=read_cols)
abs_1e3_control.columns = abs_1e3_control.columns.str.strip()
abs_1e3_control['ErrorBound'] = '1e-3'
abs_1e3_control['DecompressionBandwidth'] = (abs_1e3_control['DataSize']/abs_1e3_control['DecompressionTime'])/1048576
abs_1e3_control['IncorrectPercentage'] = (abs_1e3_control['Incorrect']/2869440)*100
abs_1e4_control = pd.read_csv(abs_control_list[2], usecols=read_cols)
abs_1e4_control.columns = abs_1e4_control.columns.str.strip()
abs_1e4_control['ErrorBound'] = '1e-4'
abs_1e4_control['DecompressionBandwidth'] = (abs_1e4_control['DataSize']/abs_1e4_control['DecompressionTime'])/1048576
abs_1e4_control['IncorrectPercentage'] = (abs_1e4_control['Incorrect']/2869440)*100
abs_control = pd.concat([abs_1e2_control, abs_1e3_control, abs_1e4_control], ignore_index=True, sort=False)
abs_control['MaxDifference'] = abs_control['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_control['Type'] = 'Control'
abs_control['Experiment'] = 'Control ABS'

abs_1e2_corrupt = pd.read_csv(abs_corrupt_list[0], usecols=read_cols)
abs_1e2_corrupt.columns = abs_1e2_corrupt.columns.str.strip()
abs_1e2_corrupt['ErrorBound'] = '1e-2'
abs_1e2_corrupt['DecompressionBandwidth'] = (abs_1e2_corrupt['DataSize']/abs_1e2_corrupt['DecompressionTime'])/1048576
abs_1e2_corrupt['IncorrectPercentage'] = (abs_1e2_corrupt['Incorrect']/2869440)*100
abs_1e2_corrupt['MaxDifference'] = abs_1e2_corrupt['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_1e3_corrupt = pd.read_csv(abs_corrupt_list[1], usecols=read_cols)
abs_1e3_corrupt.columns = abs_1e3_corrupt.columns.str.strip()
abs_1e3_corrupt['ErrorBound'] = '1e-3'
abs_1e3_corrupt['DecompressionBandwidth'] = (abs_1e3_corrupt['DataSize']/abs_1e3_corrupt['DecompressionTime'])/1048576
abs_1e3_corrupt['IncorrectPercentage'] = (abs_1e3_corrupt['Incorrect']/2869440)*100
abs_1e3_corrupt['MaxDifference'] = abs_1e3_corrupt['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_1e4_corrupt = pd.read_csv(abs_corrupt_list[2], usecols=read_cols)
abs_1e4_corrupt.columns = abs_1e4_corrupt.columns.str.strip()
abs_1e4_corrupt['ErrorBound'] = '1e-4'
abs_1e4_corrupt['DecompressionBandwidth'] = (abs_1e4_corrupt['DataSize']/abs_1e4_corrupt['DecompressionTime'])/1048576
abs_1e4_corrupt['IncorrectPercentage'] = (abs_1e4_corrupt['Incorrect']/2869440)*100
abs_1e4_corrupt['MaxDifference'] = abs_1e4_corrupt['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
abs_corrupt = pd.concat([abs_1e2_corrupt, abs_1e3_corrupt, abs_1e4_corrupt], ignore_index=True, sort=False)
abs_corrupt['Type'] = 'Corrupt'
abs_corrupt['Experiment'] = 'Corrupt ABS'

abs_total = pd.concat([abs_control, abs_corrupt], ignore_index=True, sort=False)
abs_corrupt_completed = abs_corrupt[(abs_corrupt["Status"].str.contains('Completed')) | (abs_corrupt["Status"].str.contains('Timeout'))]
```

### Control vs. Corrupt Maximum Difference

```python
sns.barplot(x='ErrorBound', y='MaxDifference', hue="Type", data=abs_total).set_title('Maximum Difference by Error Bound')
plt.yscale('log')
plt.show()
```

```python
f, axes = plt.subplots(figsize=(20, 10))
corrupt_1_pivot = abs_1e2_corrupt.pivot("FlipLocation", "ByteLocation", "MaxDifference")
fig = sns.heatmap(corrupt_1_pivot, ax=axes)
fig.set_title('ABS 1e-2: Maximum Difference By Location', fontsize="23")
fig.set_ylabel('Flip Location', fontsize="20", weight="bold")
fig.set_xlabel('Byte Location', fontsize="20", weight="bold")
cbar = fig.collections[0].colorbar
cbar.ax.tick_params(labelsize=15)
fig.set_xticks(fig.get_xticks()[::3])
fig.invert_yaxis()
f.tight_layout()
plt.show()

f, axes = plt.subplots(figsize=(20, 10))
corrupt_2_pivot = abs_1e3_corrupt.pivot("FlipLocation", "ByteLocation", "MaxDifference")
fig = sns.heatmap(corrupt_2_pivot, ax=axes)
fig.set_title('ABS 1e-3: Maximum Difference By Location', fontsize="23")
fig.set_ylabel('Flip Location', fontsize="20", weight="bold")
fig.set_xlabel('Byte Location', fontsize="20", weight="bold")
cbar = fig.collections[0].colorbar
cbar.ax.tick_params(labelsize=15)
fig.set_xticks(fig.get_xticks()[::3])
fig.invert_yaxis()
f.tight_layout()
plt.show()

f, axes = plt.subplots(figsize=(20, 10))
corrupt_3_pivot = abs_1e4_corrupt.pivot("FlipLocation", "ByteLocation", "MaxDifference")
fig = sns.heatmap(corrupt_3_pivot, ax=axes)
fig.set_title('ABS 1e-4: Maximum Difference By Location', fontsize="23")
fig.set_ylabel('Flip Location', fontsize="20", weight="bold")
fig.set_xlabel('Byte Location', fontsize="20", weight="bold")
cbar = fig.collections[0].colorbar
cbar.ax.tick_params(labelsize=15)
fig.set_xticks(fig.get_xticks()[::3])
fig.invert_yaxis()
f.tight_layout()
plt.show()
```

### Control Maximum Difference Breakdown

```python
print("ABS 1E2 Average Maximum Difference: {}".format(abs_1e2_control['MaxDifference'].mean()))
print("ABS 1E2 Maximum Difference Standard Deviation: {}\n".format(abs_1e2_control['MaxDifference'].std()))
print("ABS 1E3 Average Maximum Difference: {}".format(abs_1e3_control['MaxDifference'].mean()))
print("ABS 1E3 Maximum Difference Standard Deviation: {}\n".format(abs_1e3_control['MaxDifference'].std()))
print("ABS 1E4 Average Maximum Difference: {}".format(abs_1e4_control['MaxDifference'].mean()))
print("ABS 1E4 Maximum Difference Standard Deviation: {}\n".format(abs_1e4_control['MaxDifference'].std()))
```

### Corrupt Maximum Difference Breakdown

```python
abs_1e2_corrupt_completed = abs_1e2_corrupt[(abs_1e2_corrupt["Status"].str.contains('Completed'))]
abs_1e3_corrupt_completed = abs_1e3_corrupt[(abs_1e3_corrupt["Status"].str.contains('Completed'))]
abs_1e4_corrupt_completed = abs_1e4_corrupt[(abs_1e4_corrupt["Status"].str.contains('Completed'))]

print("ABS 1E2 Average Maximum Difference: {}".format(abs_1e2_corrupt_completed['MaxDifference'].mean()))
print("ABS 1E2 Maximum Difference Standard Deviation: {}\n".format(abs_1e2_corrupt_completed['MaxDifference'].std()))
print("ABS 1E3 Average Maximum Difference: {}".format(abs_1e3_corrupt_completed['MaxDifference'].mean()))
print("ABS 1E3 Maximum Difference Standard Deviation: {}\n".format(abs_1e3_corrupt_completed['MaxDifference'].std()))
print("ABS 1E4 Average Maximum Difference: {}".format(abs_1e4_corrupt_completed['MaxDifference'].mean()))
print("ABS 1E4 Maximum Difference Standard Deviation: {}\n".format(abs_1e4_corrupt_completed['MaxDifference'].std()))
```

When looking at these averages, we notice that there is an obvious difference between the control and corrupt cases. This difference is substantial and can also be seen in the standard deviation as well. When comparing the percentage of trials that go over the error bounding information we can also see a giant leap from the 0% in the control cases. It is also worth noting that as we increase the error bounding value the percentage beyond the error bounds goes down as well. Let's use a boxplot to see if we can see this full distribution.

```python
sns.boxplot(x='ErrorBound', y='MaxDifference', data=abs_corrupt_completed).set_title('Maximum Difference by Experiment')
plt.yscale('log')
plt.show()
```

As epected, there are many outliers that have extremely high maximum differences. This agrees with the previously seen high standard deviations seen above. Next we can look to see if an increase in maximum difference has any effect on the decompression bandwidth. To do this, we will look at the upper and lower quartiles of maximum difference and compare their average decompression bandwidths for each.

```python
abs_1e2_corrupt_upper_outliers = abs_1e2_corrupt_completed[abs_1e2_corrupt_completed['MaxDifference'] > abs_1e2_corrupt_completed.MaxDifference.quantile(0.75)]
abs_1e2_corrupt_lower_outliers = abs_1e2_corrupt_completed[abs_1e2_corrupt_completed['MaxDifference'] < abs_1e2_corrupt_completed.MaxDifference.quantile(0.25)]
abs_1e3_corrupt_upper_outliers = abs_1e3_corrupt_completed[abs_1e3_corrupt_completed['MaxDifference'] > abs_1e3_corrupt_completed.MaxDifference.quantile(0.75)]
abs_1e3_corrupt_lower_outliers = abs_1e3_corrupt_completed[abs_1e3_corrupt_completed['MaxDifference'] < abs_1e3_corrupt_completed.MaxDifference.quantile(0.25)]
abs_1e4_corrupt_upper_outliers = abs_1e4_corrupt_completed[abs_1e4_corrupt_completed['MaxDifference'] > abs_1e4_corrupt_completed.MaxDifference.quantile(0.75)]
abs_1e4_corrupt_lower_outliers = abs_1e4_corrupt_completed[abs_1e4_corrupt_completed['MaxDifference'] < abs_1e4_corrupt_completed.MaxDifference.quantile(0.25)]
```

```python
sns.scatterplot(x='MaxDifference', y='DecompressionBandwidth', data=abs_1e2_corrupt_upper_outliers).set_title('ABS 1e2: Max Diff by Decompression Bandwidth > Q3')
plt.xscale('log')
plt.show()
sns.scatterplot(x='MaxDifference', y='DecompressionBandwidth', data=abs_1e3_corrupt_upper_outliers).set_title('ABS 1e3: Max Diff by Decompression Bandwidth > Q3')
plt.xscale('log')
plt.show()
sns.scatterplot(x='MaxDifference', y='DecompressionBandwidth', data=abs_1e4_corrupt_upper_outliers).set_title('ABS 1e4: Max Diff by Decompression Bandwidth > Q3')
plt.xscale('log')
plt.show()
```

When looking at the upper quartile values it looks as though a high maximum different does not drastically change the decompression bandwidth in most cases. There are a few cases howere that do have a higher decompression bandwidth, but these seem to be less frequent and do not seem likely to indicate a pattern.

Let's now look at the lower quartile.

```python
sns.scatterplot(x='MaxDifference', y='DecompressionBandwidth', data=abs_1e2_corrupt_lower_outliers).set_title('ABS 1e2: Maximum Difference by Decompression Bandwidth < Q1')
plt.show()
sns.scatterplot(x='MaxDifference', y='DecompressionBandwidth', data=abs_1e3_corrupt_lower_outliers).set_title('ABS 1e3: Maximum Difference by Decompression Bandwidth < Q1')
plt.show()
sns.scatterplot(x='MaxDifference', y='DecompressionBandwidth', data=abs_1e4_corrupt_lower_outliers).set_title('ABS 1e4: Maximum Difference by Decompression Bandwidth < Q1')
plt.show()
```

Unlike the upper quartile case, the decompression bandwidth varies drastically when looking at items lower than the first quartile. This is interesting and could be something more.

Let's see if we can pin point some bit flip locations that lead to the largest Maximum Differences

```python
sns.scatterplot(x='ByteLocation', y='MaxDifference', data=abs_1e2_corrupt_completed).set_title('ABS 1e2: Byte Location by Maximum Difference')
plt.xscale('log')
plt.yscale('log')
plt.ylim(10, 1000000)
plt.show()
```

```python
sns.scatterplot(x='ByteLocation', y='MaxDifference', data=abs_1e3_corrupt_completed).set_title('ABS 1e3: Byte Location by Maximum Difference')
plt.xscale('log')
plt.yscale('log')
plt.ylim(10, 1000000)
plt.show()
```

```python
sns.scatterplot(x='ByteLocation', y='MaxDifference', data=abs_1e4_corrupt_completed).set_title('ABS 1e4: Byte Location by Maximum Difference')
plt.xscale('log')
plt.yscale('log')
plt.ylim(10, 1000000)
plt.show()
```

From the above information we can see that there is most likely a group of bytes that can cause drastic jumps in maximum difference if hit. This group would most likely include bytes 44, 3, 58, and 67 among others. It would be interesting to see how protecting these bits would reduce the maximum difference.

```python
f, axes = plt.subplots(figsize=(12, 7))
fig = sns.kdeplot(abs_1e2_corrupt_upper_outliers['ByteLocation'], shade=True, label='1e-2')
sns.kdeplot(abs_1e3_corrupt_upper_outliers['ByteLocation'], shade=True, label='1e-3')
sns.kdeplot(abs_1e4_corrupt_upper_outliers['ByteLocation'], shade=True, label='1e-4')
fig.set_xlabel('Byte Location', fontsize=14, weight='bold')
fig.set_ylabel('Probability', fontsize=14, weight='bold')
fig.set_title('ABS: Maximum Difference > Q3', fontsize=20)
plt.legend(title='ErrorBound')
plt.setp(fig.get_legend().get_texts(), fontsize='13')
plt.setp(fig.get_legend().get_title(), fontsize='15')
plt.tight_layout()
plt.xscale('log')
plt.show()

f, axes = plt.subplots(figsize=(12, 7))
fig = sns.kdeplot(abs_1e2_corrupt_lower_outliers['ByteLocation'], shade=True, label='1e-2')
sns.kdeplot(abs_1e3_corrupt_lower_outliers['ByteLocation'], shade=True, label='1e-3')
sns.kdeplot(abs_1e4_corrupt_lower_outliers['ByteLocation'], shade=True, label='1e-4')
fig.set_xlabel('Byte Location', fontsize=14, weight='bold')
fig.set_ylabel('Probability', fontsize=14, weight='bold')
fig.set_title('ABS: Maximum Difference < Q1', fontsize=20)
plt.legend(title='ErrorBound')
plt.setp(fig.get_legend().get_texts(), fontsize='13')
plt.setp(fig.get_legend().get_title(), fontsize='15')
plt.tight_layout()
plt.xscale('log')
plt.show()
```

```python
psnr_60_control = pd.read_csv(psnr_control_list[0], usecols=read_cols)
psnr_60_control.columns = psnr_60_control.columns.str.strip()
psnr_60_control['ErrorBound'] = '060'
psnr_60_control['DecompressionBandwidth'] = (psnr_60_control['DataSize']/psnr_60_control['DecompressionTime'])/1048576
psnr_90_control = pd.read_csv(psnr_control_list[1], usecols=read_cols)
psnr_90_control.columns = psnr_90_control.columns.str.strip()
psnr_90_control['ErrorBound'] = '090'
psnr_90_control['DecompressionBandwidth'] = (psnr_90_control['DataSize']/psnr_90_control['DecompressionTime'])/1048576
psnr_120_control = pd.read_csv(psnr_control_list[2], usecols=read_cols)
psnr_120_control.columns = psnr_120_control.columns.str.strip()
psnr_120_control['ErrorBound'] = '120'
psnr_120_control['DecompressionBandwidth'] = (psnr_120_control['DataSize']/psnr_120_control['DecompressionTime'])/1048576
psnr_control = pd.concat([psnr_60_control, psnr_90_control, psnr_120_control], ignore_index=True, sort=False)
psnr_control['MaxDifference'] = psnr_control['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
psnr_control['Type'] = 'Control'
psnr_control['Experiment'] = 'Control PSNR'

psnr_60_corrupt = pd.read_csv(psnr_corrupt_list[0], usecols=read_cols)
psnr_60_corrupt.columns = psnr_60_corrupt.columns.str.strip()
psnr_60_corrupt['ErrorBound'] = '060'
psnr_60_corrupt['DecompressionBandwidth'] = (psnr_60_corrupt['DataSize']/psnr_60_corrupt['DecompressionTime'])/1048576
psnr_60_corrupt['MaxDifference'] = psnr_60_corrupt['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
psnr_90_corrupt = pd.read_csv(psnr_corrupt_list[1], usecols=read_cols)
psnr_90_corrupt.columns = psnr_90_corrupt.columns.str.strip()
psnr_90_corrupt['ErrorBound'] = '090'
psnr_90_corrupt['DecompressionBandwidth'] = (psnr_90_corrupt['DataSize']/psnr_90_corrupt['DecompressionTime'])/1048576
psnr_90_corrupt['MaxDifference'] = psnr_90_corrupt['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
psnr_120_corrupt = pd.read_csv(psnr_corrupt_list[2], usecols=read_cols)
psnr_120_corrupt.columns = psnr_120_corrupt.columns.str.strip()
psnr_120_corrupt['ErrorBound'] = '120'
psnr_120_corrupt['DecompressionBandwidth'] = (psnr_120_corrupt['DataSize']/psnr_120_corrupt['DecompressionTime'])/1048576
psnr_120_corrupt['MaxDifference'] = psnr_120_corrupt['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
psnr_corrupt = pd.concat([psnr_60_corrupt, psnr_90_corrupt, psnr_120_corrupt], ignore_index=True, sort=False)
psnr_corrupt['Type'] = 'Corrupt'
psnr_corrupt['Experiment'] = 'Corrupt PSNR'

psnr_total = pd.concat([psnr_control, psnr_corrupt], ignore_index=True, sort=False)
psnr_corrupt_completed = psnr_corrupt[(psnr_corrupt["Status"].str.contains('Completed')) | (psnr_corrupt["Status"].str.contains('Timeout'))]
```

## Answers and Conclusions:

In the above analysis we looked at answering the following questions and from our analysis we are able to answer these questions better than before:

1. What is the average and std of maximum difference on control and corrupt cases?
  
The average and std of maximum difference in the control case is relatively small on average with very little variance. However, when we added in faults, the maximum difference skyrockets along with the standard deviation.

2. What percentage of trials violate the error bound by maximum difference?
  
Due to the fact that we are not able to accurately determine the exact error bound for PW and PSNR, we are only able to determine this percentage for ABS. When looking at ABS, we noticed that injecting a fault will violate the error bounds 77-99% of the time for error bounds between 1e4 and 1e2.

3. Which bit flips cause the largest maximum difference in the decompressed results?
  
When looking at each error bounding mode, we were able to see that giant spikes in maximum difference occur towards the earlier bytes in the compressed data with some giant spikes also occuring at smaller error bounding values. It is also worth noting that PW had the highest differences among all.

4. Does an increased maximum difference decrease decompression bandwidth?
  
Overall, decompression bandwidth is effected by maximum difference, but not by as much as one would expect. For instance larger maximum differences did not lead to a lower decompression bandwidth and smaller maximum differences caused larger variations in decompression bandwidths but not on a large scale.
