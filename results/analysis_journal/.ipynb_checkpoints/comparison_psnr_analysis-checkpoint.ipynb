# EXAALT PSNR Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the psnr changes with each of the other metrics.

In particular, we will be aiming to answer the following questions:
1. What is the average and standard deviation of PSNR in control and corrupt cases?
2. How does corruption effect PSNR?
3. Does a higher PSNR correlate with a change in Decompression Bandwidth?
4. Which bit flips cause the highest PSNR?

## ABS Analysis

### Read in data

```python
A_control_list = A_abs_control_list
A_corrupt_list = A_abs_corrupt_list

B_control_list = B_abs_control_list
B_corrupt_list = B_abs_corrupt_list

A_error_bounds = ['1e-2', '1e-3']
B_error_bounds = ['1e-2', '1e-3']

A_control_1 = pd.read_csv(A_control_list[0], usecols=read_cols)
A_control_1.columns = A_control_1.columns.str.strip()
A_control_1['ErrorBound'] = A_error_bounds[0]
A_control_1['DecompressionBandwidth'] = (A_control_1['DataSize']/A_control_1['DecompressionTime'])/1048576
A_control_1['PSNR'] = A_control_1['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
A_control_2 = pd.read_csv(A_control_list[1], usecols=read_cols)
A_control_2.columns = A_control_2.columns.str.strip()
A_control_2['ErrorBound'] = A_error_bounds[1]
A_control_2['DecompressionBandwidth'] = (A_control_2['DataSize']/A_control_2['DecompressionTime'])/1048576
A_control_2['PSNR'] = A_control_2['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
A_control = pd.concat([A_control_1, A_control_2], ignore_index=True, sort=False)
A_control['Experiment'] = 'EXAALT Control ABS'

A_corrupt_1 = pd.read_csv(A_corrupt_list[0], usecols=read_cols)
A_corrupt_1.columns = A_corrupt_1.columns.str.strip()
A_corrupt_1['ErrorBound'] = A_error_bounds[0]
A_corrupt_1['DecompressionBandwidth'] = (A_corrupt_1['DataSize']/A_corrupt_1['DecompressionTime'])/1048576
A_corrupt_1['PSNR'] = A_corrupt_1['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
A_corrupt_2 = pd.read_csv(A_corrupt_list[1], usecols=read_cols)
A_corrupt_2.columns = A_corrupt_2.columns.str.strip()
A_corrupt_2['ErrorBound'] = A_error_bounds[1]
A_corrupt_2['DecompressionBandwidth'] = (A_corrupt_2['DataSize']/A_corrupt_2['DecompressionTime'])/1048576
A_corrupt_2['PSNR'] = A_corrupt_2['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
A_corrupt = pd.concat([A_corrupt_1, A_corrupt_2], ignore_index=True, sort=False)
A_corrupt['Experiment'] = 'EXAALT Corrupt ABS'

A_total = pd.concat([A_control, A_corrupt], ignore_index=True, sort=False)

B_control_1 = pd.read_csv(B_control_list[0], usecols=read_cols)
B_control_1.columns = B_control_1.columns.str.strip()
B_control_1['ErrorBound'] = B_error_bounds[0]
B_control_1['DecompressionBandwidth'] = (B_control_1['DataSize']/B_control_1['DecompressionTime'])/1048576
B_control_1['PSNR'] = B_control_1['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
B_control_2 = pd.read_csv(B_control_list[1], usecols=read_cols)
B_control_2.columns = B_control_2.columns.str.strip()
B_control_2['ErrorBound'] = B_error_bounds[1]
B_control_2['DecompressionBandwidth'] = (B_control_2['DataSize']/B_control_2['DecompressionTime'])/1048576
B_control_2['PSNR'] = B_control_2['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)

B_control = pd.concat([B_control_1, B_control_2], ignore_index=True, sort=False)
B_control['Experiment'] = 'HACC Control ABS'

B_corrupt_1 = pd.read_csv(B_corrupt_list[0], usecols=read_cols)
B_corrupt_1.columns = B_corrupt_1.columns.str.strip()
B_corrupt_1['ErrorBound'] = B_error_bounds[0]
B_corrupt_1['DecompressionBandwidth'] = (B_corrupt_1['DataSize']/B_corrupt_1['DecompressionTime'])/1048576
B_corrupt_1['PSNR'] = B_corrupt_1['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
B_corrupt_2 = pd.read_csv(B_corrupt_list[1], usecols=read_cols)
B_corrupt_2.columns = B_corrupt_2.columns.str.strip()
B_corrupt_2['ErrorBound'] = B_error_bounds[1]
B_corrupt_2['DecompressionBandwidth'] = (B_corrupt_2['DataSize']/B_corrupt_2['DecompressionTime'])/1048576
B_corrupt_2['PSNR'] = B_corrupt_2['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
B_corrupt = pd.concat([B_corrupt_1, B_corrupt_2], ignore_index=True, sort=False)
B_corrupt['Experiment'] = 'HACC Corrupt ABS'

B_total = pd.concat([B_control, B_corrupt], ignore_index=True, sort=False)

abs_total = pd.concat([A_total, B_total], ignore_index=True, sort=False)
abs_completed = abs_total[(abs_total["Status"].str.contains('Completed')) | (abs_total["Status"].str.contains('Timeout'))]
abs_corrupt = pd.concat([A_corrupt, B_corrupt], ignore_index=True, sort=False)

exaalt_abs_1e2_corrupt_completed = A_corrupt_1[(A_corrupt_1["Status"].str.contains('Completed'))]
exaalt_abs_1e3_corrupt_completed = A_corrupt_2[(A_corrupt_2["Status"].str.contains('Completed'))]

hacc_abs_1e2_corrupt_completed = B_corrupt_1[(B_corrupt_1["Status"].str.contains('Completed'))]
hacc_abs_1e3_corrupt_completed = B_corrupt_2[(B_corrupt_2["Status"].str.contains('Completed'))]
```

## PW Analysis

### Read in data

```python
control_list = pw_control_list
corrupt_list = pw_corrupt_list
error_bounds = ['1e-2', '1e-3', '1e-4']

control_1 = pd.read_csv(control_list[0], usecols=read_cols)
control_1.columns = control_1.columns.str.strip()
control_1['ErrorBound'] = error_bounds[0]
control_1['DecompressionBandwidth'] = (control_1['DataSize']/control_1['DecompressionTime'])/1048576
control_1['PSNR'] = control_1['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_2 = pd.read_csv(control_list[1], usecols=read_cols)
control_2.columns = control_2.columns.str.strip()
control_2['ErrorBound'] = error_bounds[1]
control_2['DecompressionBandwidth'] = (control_2['DataSize']/control_2['DecompressionTime'])/1048576
control_2['PSNR'] = control_2['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_3 = pd.read_csv(control_list[2], usecols=read_cols)
control_3.columns = control_3.columns.str.strip()
control_3['ErrorBound'] = error_bounds[2]
control_3['DecompressionBandwidth'] = (control_3['DataSize']/control_3['DecompressionTime'])/1048576
control_3['PSNR'] = control_3['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control = pd.concat([control_1, control_2, control_3], ignore_index=True, sort=False)
control['Type'] = 'Control'
control['Experiment'] = 'Control PW-REL'

corrupt_1 = pd.read_csv(corrupt_list[0], usecols=read_cols)
corrupt_1.columns = corrupt_1.columns.str.strip()
corrupt_1['ErrorBound'] = error_bounds[0]
corrupt_1['DecompressionBandwidth'] = (corrupt_1['DataSize']/corrupt_1['DecompressionTime'])/1048576
corrupt_1['PSNR'] = corrupt_1['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_2 = pd.read_csv(corrupt_list[1], usecols=read_cols)
corrupt_2.columns = corrupt_2.columns.str.strip()
corrupt_2['ErrorBound'] = error_bounds[1]
corrupt_2['DecompressionBandwidth'] = (corrupt_2['DataSize']/corrupt_2['DecompressionTime'])/1048576
corrupt_2['PSNR'] = corrupt_2['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_3 = pd.read_csv(corrupt_list[2], usecols=read_cols)
corrupt_3.columns = corrupt_3.columns.str.strip()
corrupt_3['ErrorBound'] = error_bounds[2]
corrupt_3['DecompressionBandwidth'] = (corrupt_3['DataSize']/corrupt_3['DecompressionTime'])/1048576
corrupt_3['PSNR'] = corrupt_3['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt = pd.concat([corrupt_1, corrupt_2, corrupt_3], ignore_index=True, sort=False)
corrupt['Type'] = 'Corrupt'
corrupt['Experiment'] = 'Corrupt PW-REL'

total = pd.concat([control, corrupt], ignore_index=True, sort=False)
pw_total = total.copy()
corrupt_completed = corrupt[(corrupt["Status"].str.contains('Completed')) | (corrupt["Status"].str.contains('Timeout'))]
```

### Corrupt Maximum Difference Breakdown

```python
corrupt_1_completed = corrupt_1[(corrupt_1["Status"].str.contains('Completed'))]
corrupt_2_completed = corrupt_2[(corrupt_2["Status"].str.contains('Completed'))]
corrupt_3_completed = corrupt_3[(corrupt_3["Status"].str.contains('Completed'))]

print("PW 1E2 Average PSNR: {}".format(corrupt_1_completed['PSNR'].mean()))
print("PW 1E2 PSNR Standard Deviation: {}\n".format(corrupt_1_completed['PSNR'].std()))
print("PW 1E3 Average PSNR: {}".format(corrupt_2_completed['PSNR'].mean()))
print("PW 1E3 PSNR Standard Deviation: {}\n".format(corrupt_2_completed['PSNR'].std()))
print("PW 1E4 Average PSNR: {}".format(corrupt_3_completed['PSNR'].mean()))
print("PW 1E4 PSNR Standard Deviation: {}\n".format(corrupt_3_completed['PSNR'].std()))
```

Looking at these averages, we notice that there is a minor jump in the number of incorrect values between the control and corrupt cases. The standard deviation is much higher though leading to a high variance in the number of incorrect values that can be seen. Let's use a boxplot to see if we can see the full distribution.

```python
sns.boxplot(x='ErrorBound', y='PSNR', data=corrupt_completed).set_title('PSNR by Error Bound')
plt.yscale('log')
plt.show()
```

As expected, there are a ton of outliers which explains the high standard deviation. Therefore, in most cases only a small percentage of values will be incorrect, but there could be a ton of them as well. We can look to see if an increase in incorrect percentages has any effect on the decompression bandwidth.

```python
f, axes = plt.subplots(1, 2, figsize=(20, 7))
sns.scatterplot(x='PSNR', y='DecompressionBandwidth', data=corrupt_1_completed, ax=axes[0]).set_title('PW 1e2: PSNR by Decompression Bandwidth')
sns.scatterplot(x='PSNR', y='DecompressionBandwidth', data=corrupt_1_completed, ax=axes[1]).set_title('PW 1e2: PSNR by Decompression Bandwidth')
plt.yscale('log')
plt.show()

f, axes = plt.subplots(1, 2, figsize=(20, 7))
sns.scatterplot(x='PSNR', y='DecompressionBandwidth', data=corrupt_2_completed, ax=axes[0]).set_title('PW 1e3: PSNR by Decompression Bandwidth')
sns.scatterplot(x='PSNR', y='DecompressionBandwidth', data=corrupt_2_completed, ax=axes[1]).set_title('PW 1e3: PSNR by Decompression Bandwidth')
plt.yscale('log')
plt.show()

f, axes = plt.subplots(1, 2, figsize=(20, 7))
sns.scatterplot(x='PSNR', y='DecompressionBandwidth', data=corrupt_3_completed, ax=axes[0]).set_title('PW 1e4: PSNR by Decompression Bandwidth')
sns.scatterplot(x='PSNR', y='DecompressionBandwidth', data=corrupt_3_completed, ax=axes[1]).set_title('PW 1e4: PSNR by Decompression Bandwidth')
plt.yscale('log')
plt.show()
```

Looking at these results, it seems that the decompression bandwidth effects are centered in two main locations.

Looking at the results above we can see that there are two major distributions of PSNR when compared to Incorrect Percentage. Looking closer we can see the patterns of such.

Let's see if we can pin point some bit flip locations that lead to the largest Incorrect Percentage

```python
sns.scatterplot(x='ByteLocation', y='PSNR', data=corrupt_1_completed).set_title('PW 1e2: Byte Location by PSNR')
plt.xscale('log')
plt.show()
```

```python
sns.scatterplot(x='ByteLocation', y='PSNR', data=corrupt_2_completed).set_title('PW 1e3: Byte Location by PSNR')
plt.xscale('log')
plt.show()
```

```python
sns.scatterplot(x='ByteLocation', y='PSNR', data=corrupt_3_completed).set_title('PW 1e4: Byte Location by PSNR')
plt.xscale('log')
plt.show()
```

```python
pw_1e2_corrupt_upper_outliers = corrupt_1_completed[corrupt_1_completed['PSNR'] > corrupt_1_completed.PSNR.quantile(0.75)]
pw_1e2_corrupt_lower_outliers = corrupt_1_completed[corrupt_1_completed['PSNR'] < corrupt_1_completed.PSNR.quantile(0.25)]
pw_1e3_corrupt_upper_outliers = corrupt_2_completed[corrupt_2_completed['PSNR'] > corrupt_2_completed.PSNR.quantile(0.75)]
pw_1e3_corrupt_lower_outliers = corrupt_2_completed[corrupt_2_completed['PSNR'] < corrupt_2_completed.PSNR.quantile(0.25)]
pw_1e4_corrupt_upper_outliers = corrupt_3_completed[corrupt_3_completed['PSNR'] > corrupt_3_completed.PSNR.quantile(0.75)]
pw_1e4_corrupt_lower_outliers = corrupt_3_completed[corrupt_3_completed['PSNR'] < corrupt_3_completed.PSNR.quantile(0.25)]

f, axes = plt.subplots(figsize=(12, 7))
fig = sns.kdeplot(pw_1e2_corrupt_upper_outliers['ByteLocation'], shade=True, label='1e-2')
sns.kdeplot(pw_1e3_corrupt_upper_outliers['ByteLocation'], shade=True, label='1e-3')
sns.kdeplot(pw_1e4_corrupt_upper_outliers['ByteLocation'], shade=True, label='1e-4')
fig.set_xlabel('Byte Location', fontsize=14, weight='bold')
fig.set_ylabel('Probability', fontsize=14, weight='bold')
fig.set_title('PW-REL: PSNR > Q3', fontsize=20)
plt.legend(title='ErrorBound')
plt.setp(fig.get_legend().get_texts(), fontsize='13')
plt.setp(fig.get_legend().get_title(), fontsize='15')
plt.tight_layout()
plt.xscale('log')
plt.show()

f, axes = plt.subplots(figsize=(12, 7))
fig = sns.kdeplot(pw_1e2_corrupt_lower_outliers['ByteLocation'], shade=True, label='1e-2')
sns.kdeplot(pw_1e3_corrupt_lower_outliers['ByteLocation'], shade=True, label='1e-3')
sns.kdeplot(pw_1e4_corrupt_lower_outliers['ByteLocation'], shade=True, label='1e-4')
fig.set_xlabel('Byte Location', fontsize=14, weight='bold')
fig.set_ylabel('Probability', fontsize=14, weight='bold')
fig.set_title('PW-REL: PSNR < Q1', fontsize=20)
plt.legend(title='ErrorBound')
plt.setp(fig.get_legend().get_texts(), fontsize='13')
plt.setp(fig.get_legend().get_title(), fontsize='15')
plt.tight_layout()
plt.xscale('log')
plt.show()
```
