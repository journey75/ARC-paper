# EXAALT Incorrect Element Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the incorrect percentage changes with each of the other metrics.

In particular, we will be aiming to answer the following questions:
1. What is the average and std number of incorrect elements on control and corrupt cases?
2. How does incorrect count correlate with the other metrics?
3. Which bit flips cause the largest percentage of incorrect count?
4. What percentage of bit flips cause incorrect values?
4. Does an increase in incorrect elements decrease decompression bandwidth?

## Libraries and Constants

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib
import seaborn as sns
import time
import numpy as np

sns.set()
sns.set_style("whitegrid", {'axes.grid' : True})
sns.set_color_codes()
sns.set_context("notebook", font_scale=1.1, rc={"lines.linewidth" : 3.5, "lines.markeredgewidth" : 1., "line.markersize" : 10})
font = {"family" : 'serif'} 
matplotlib.rc('font', **font)
color = ["b", "g", "r", "y", "m"]

abs_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_abs_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_control_abs_1e4.csv']
abs_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_abs_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_abs_1e4.csv']

pw_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_pw_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_control_pw_1e4.csv']
pw_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_pw_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_pw_1e4.csv']

psnr_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_psnr_60.csv', './notebook_data/exaalt_data/sz_exaalt_control_psnr_90.csv', './notebook_data/exaalt_data/sz_exaalt_control_psnr_120.csv']
psnr_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_psnr_60.csv', './notebook_data/exaalt_data/sz_exaalt_psnr_90.csv', './notebook_data/exaalt_data/sz_exaalt_psnr_120.csv']

read_cols=["DataSize", "ByteLocation", "FlipLocation", "DecompressionTime", "Incorrect", "MaxDifference", "RMSE", "PSNR", "Status"]
data_elements = 2869440
```

## ABS Analysis

### Read in data

```python
control_list = abs_control_list
corrupt_list = abs_corrupt_list
error_bounds = ['1e-2', '1e-3', '1e-4']

control_1 = pd.read_csv(control_list[0], usecols=read_cols)
control_1.columns = control_1.columns.str.strip()
control_1['ErrorBound'] = error_bounds[0]
control_1['DecompressionBandwidth'] = (control_1['DataSize']/control_1['DecompressionTime'])/1048576
control_1['IncorrectPercentage'] = (control_1['Incorrect']/data_elements)*100
control_1['MaxDifference'] = control_1['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_1['RMSE'] = control_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_1['PSNR'] = control_1['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_2 = pd.read_csv(control_list[1], usecols=read_cols)
control_2.columns = control_2.columns.str.strip()
control_2['ErrorBound'] = error_bounds[1]
control_2['DecompressionBandwidth'] = (control_2['DataSize']/control_2['DecompressionTime'])/1048576
control_2['IncorrectPercentage'] = (control_2['Incorrect']/data_elements)*100
control_2['MaxDifference'] = control_2['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_2['RMSE'] = control_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_2['PSNR'] = control_2['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_3 = pd.read_csv(control_list[2], usecols=read_cols)
control_3.columns = control_3.columns.str.strip()
control_3['ErrorBound'] = error_bounds[2]
control_3['DecompressionBandwidth'] = (control_3['DataSize']/control_3['DecompressionTime'])/1048576
control_3['IncorrectPercentage'] = (control_3['Incorrect']/data_elements)*100
control_3['MaxDifference'] = control_3['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_3['RMSE'] = control_3['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_3['PSNR'] = control_3['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control = pd.concat([control_1, control_2, control_3], ignore_index=True, sort=False)
control['Type'] = 'Control'

corrupt_1 = pd.read_csv(corrupt_list[0], usecols=read_cols)
corrupt_1.columns = corrupt_1.columns.str.strip()
corrupt_1['ErrorBound'] = error_bounds[0]
corrupt_1['DecompressionBandwidth'] = (corrupt_1['DataSize']/corrupt_1['DecompressionTime'])/1048576
corrupt_1['IncorrectPercentage'] = (corrupt_1['Incorrect']/data_elements)*100
corrupt_1['MaxDifference'] = corrupt_1['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_1['RMSE'] = corrupt_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_1['PSNR'] = corrupt_1['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_2 = pd.read_csv(corrupt_list[1], usecols=read_cols)
corrupt_2.columns = corrupt_2.columns.str.strip()
corrupt_2['ErrorBound'] = error_bounds[1]
corrupt_2['DecompressionBandwidth'] = (corrupt_2['DataSize']/corrupt_2['DecompressionTime'])/1048576
corrupt_2['IncorrectPercentage'] = (corrupt_2['Incorrect']/data_elements)*100
corrupt_2['MaxDifference'] = corrupt_2['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_2['RMSE'] = corrupt_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_2['PSNR'] = corrupt_2['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_3 = pd.read_csv(corrupt_list[2], usecols=read_cols)
corrupt_3.columns = corrupt_3.columns.str.strip()
corrupt_3['ErrorBound'] = error_bounds[2]
corrupt_3['DecompressionBandwidth'] = (corrupt_3['DataSize']/corrupt_3['DecompressionTime'])/1048576
corrupt_3['IncorrectPercentage'] = (corrupt_3['Incorrect']/data_elements)*100
corrupt_3['MaxDifference'] = corrupt_3['MaxDifference'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_3['RMSE'] = corrupt_3['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_3['PSNR'] = corrupt_3['PSNR'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt = pd.concat([corrupt_1, corrupt_2, corrupt_3], ignore_index=True, sort=False)
corrupt['Type'] = 'Corrupt'
corrupt['Experiment'] = 'ABS'

total = pd.concat([control, corrupt], ignore_index=True, sort=False)
corrupt_completed = corrupt[(corrupt["Status"].str.contains('Completed')) | (corrupt["Status"].str.contains('Timeout'))]
```

### Control vs. Corrupt Incorrect Percentages

```python
sns.barplot(x='ErrorBound', y='IncorrectPercentage', data=corrupt, ci=None).set_title('ABS Incorrect Elements by Error Bound', fontsize=15, weight='bold')
plt.xlabel("Error Bound",fontsize=14, weight='bold')
plt.ylabel("Incorrect Element (%)",fontsize=14,weight='bold')
plt.tight_layout()
plt.show()
```

```python
f, axes = plt.subplots(figsize=(20, 10))
corrupt_1_pivot = corrupt_1.pivot("FlipLocation", "ByteLocation", "Incorrect")
fig = sns.heatmap(corrupt_1_pivot, ax=axes)
fig.set_title('ABS 1e-2: Incorrect Element Count By Location', fontsize="23")
fig.set_ylabel('Flip Location', fontsize="20", weight="bold")
fig.set_xlabel('Byte Location', fontsize="20", weight="bold")
cbar = fig.collections[0].colorbar
cbar.ax.tick_params(labelsize=15)
fig.set_xticks(fig.get_xticks()[::3])
fig.invert_yaxis()
f.tight_layout()
plt.show()

f, axes = plt.subplots(figsize=(20, 10))
corrupt_2_pivot = corrupt_2.pivot("FlipLocation", "ByteLocation", "Incorrect")
fig = sns.heatmap(corrupt_2_pivot, ax=axes)
fig.set_title('ABS 1e-3: Incorrect Element Count By Location', fontsize="23")
fig.set_ylabel('Flip Location', fontsize="20", weight="bold")
fig.set_xlabel('Byte Location', fontsize="20", weight="bold")
cbar = fig.collections[0].colorbar
cbar.ax.tick_params(labelsize=15)
fig.set_xticks(fig.get_xticks()[::3])
fig.invert_yaxis()
f.tight_layout()
plt.show()

f, axes = plt.subplots(figsize=(20, 10))
corrupt_3_pivot = corrupt_3.pivot("FlipLocation", "ByteLocation", "Incorrect")
fig = sns.heatmap(corrupt_3_pivot, ax=axes)
fig.set_title('ABS 1e-4: Incorrect Element Count By Location', fontsize="23")
fig.set_ylabel('Flip Location', fontsize="20", weight="bold")
fig.set_xlabel('Byte Location', fontsize="20", weight="bold")
cbar = fig.collections[0].colorbar
cbar.ax.tick_params(labelsize=15)
fig.set_xticks(fig.get_xticks()[::3])
fig.invert_yaxis()
f.tight_layout()
plt.show()
```

### Corrupt Incorrect Breakdown

```python
corrupt_1_completed = corrupt_1[(corrupt_1["Status"].str.contains('Completed'))]
corrupt_1_completed = corrupt_1_completed.copy()
corrupt_1_incorrects = corrupt_1_completed[corrupt_1_completed['IncorrectPercentage'] > 0]
corrupt_2_completed = corrupt_2[(corrupt_2["Status"].str.contains('Completed'))]
corrupt_2_completed = corrupt_2_completed.copy()
corrupt_2_incorrects = corrupt_2_completed[corrupt_2_completed['IncorrectPercentage'] > 0]
corrupt_3_completed = corrupt_3[(corrupt_3["Status"].str.contains('Completed'))]
corrupt_3_completed = corrupt_3_completed.copy()
corrupt_3_incorrects = corrupt_3_completed[corrupt_3_completed['IncorrectPercentage'] > 0]

print("ABS 1E2 Average Incorrect Percentage: {}%".format(corrupt_1_completed['IncorrectPercentage'].mean()))
print("ABS 1E2 Incorrect Percentage Standard Deviation: {}%".format(corrupt_1_completed['IncorrectPercentage'].std()))
print("ABS 1E2 Percent of Trials with Incorrect Elements: {}%\n".format(corrupt_1_incorrects.shape[0]/data_elements*100))
print("ABS 1E3 Average Incorrect Percentage: {}%".format(corrupt_2_completed['IncorrectPercentage'].mean()))
print("ABS 1E3 Incorrect Percentage Standard Deviation: {}%".format(corrupt_2_completed['IncorrectPercentage'].std()))
print("ABS 1E3 Percent of Trials with Incorrect Elements: {}%\n".format(corrupt_2_incorrects.shape[0]/data_elements*100))
print("ABS 1E4 Average Incorrect Percentage: {}%".format(corrupt_3_completed['IncorrectPercentage'].mean()))
print("ABS 1E4 Incorrect Percentage Standard Deviation: {}%".format(corrupt_3_completed['IncorrectPercentage'].std()))
print("ABS 1E4 Percent of Trials with Incorrect Elements: {}%\n".format(corrupt_3_incorrects.shape[0]/data_elements*100))
```

Looking at these averages, we notice that there is a minor jump in the number of incorrect values between the control and corrupt cases. The standard deviation is much higher though leading to a high variance in the number of incorrect values that can be seen. Let's use a boxplot to see if we can see the full distribution.

```python
sns.boxplot(x='ErrorBound', y='IncorrectPercentage', data=corrupt_completed).set_title('Incorrect Percentage by Error Bound')
plt.show()
```

As expected, there are a ton of outliers which explains the high standard deviation. Therefore, in most cases only a small percentage of values will be incorrect, but there could be a ton of them as well. We can look to see if an increase in incorrect percentages has any effect on the decompression bandwidth.

```python
sns.scatterplot(x='IncorrectPercentage', y='DecompressionBandwidth', data=corrupt_1_completed).set_title('ABS 1e2: Incorrect by Decompression Bandwidth')
plt.yscale('log')
plt.show()
sns.scatterplot(x='IncorrectPercentage', y='DecompressionBandwidth', data=corrupt_2_completed).set_title('ABS 1e3: Incorrect by Decompression Bandwidth')
plt.yscale('log')
plt.show()
sns.scatterplot(x='IncorrectPercentage', y='DecompressionBandwidth', data=corrupt_3_completed).set_title('ABS 1e4: Incorrect by Decompression Bandwidth')
plt.yscale('log')
plt.show()
```
