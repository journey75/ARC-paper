# EXAALT RMSE Analysis

This analysis is part of the overall EXAALT experiment analysis. In this section we will be looking at how the rmse changes with each of the other metrics.

In particular, we will be aiming to answer the following questions:
1. What is the average and standard deviation of RMSE in control and corrupt cases?
2. How does corruption effect RMSE?
3. Does a higher RMSE correlate with a change in Decompression Bandwidth?
4. Which bit flips cause the highest RMSE?

## Libraries and Constants

```python
import sys
import os
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
import seaborn as sns
import time

sns.set()
sns.set_style("whitegrid", {'axes.grid' : True})
sns.set_color_codes()
sns.set_context("notebook", font_scale=1.1, rc={"lines.linewidth" : 3.5, "lines.markeredgewidth" : 1., "line.markersize" : 10})
font = {"family" : 'serif'} 
matplotlib.rc('font', **font)
color = ["b", "g", "r", "y", "m"]

abs_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_abs_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_control_abs_1e4.csv']
abs_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_abs_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_abs_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_abs_1e4.csv']

pw_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_control_pw_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_control_pw_1e4.csv']
pw_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_pw_1e2.csv', './notebook_data/exaalt_data/sz_exaalt_pw_1e3.csv', './notebook_data/exaalt_data/sz_exaalt_pw_1e4.csv']

psnr_control_list = ['./notebook_data/exaalt_data/sz_exaalt_control_psnr_60.csv', './notebook_data/exaalt_data/sz_exaalt_control_psnr_90.csv', './notebook_data/exaalt_data/sz_exaalt_control_psnr_120.csv']
psnr_corrupt_list = ['./notebook_data/exaalt_data/sz_exaalt_psnr_60.csv', './notebook_data/exaalt_data/sz_exaalt_psnr_90.csv', './notebook_data/exaalt_data/sz_exaalt_psnr_120.csv']

read_cols=["DataSize", "ByteLocation", "FlipLocation", "DecompressionTime", "Incorrect", "RMSE", "Status"]
data_elements = 2869440
```

## ABS Analysis

### Read in data

```python
control_list = abs_control_list
corrupt_list = abs_corrupt_list
error_bounds = ['1e-2', '1e-3', '1e-4']

control_1 = pd.read_csv(control_list[0], usecols=read_cols)
control_1.columns = control_1.columns.str.strip()
control_1['ErrorBound'] = error_bounds[0]
control_1['DecompressionBandwidth'] = (control_1['DataSize']/control_1['DecompressionTime'])/1048576
control_1['RMSE'] = control_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_2 = pd.read_csv(control_list[1], usecols=read_cols)
control_2.columns = control_2.columns.str.strip()
control_2['ErrorBound'] = error_bounds[1]
control_2['DecompressionBandwidth'] = (control_2['DataSize']/control_2['DecompressionTime'])/1048576
control_2['RMSE'] = control_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_3 = pd.read_csv(control_list[2], usecols=read_cols)
control_3.columns = control_3.columns.str.strip()
control_3['ErrorBound'] = error_bounds[2]
control_3['DecompressionBandwidth'] = (control_3['DataSize']/control_3['DecompressionTime'])/1048576
control_3['RMSE'] = control_3['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control = pd.concat([control_1, control_2, control_3], ignore_index=True, sort=False)
control['Type'] = 'Control'
control['Experiment'] = 'Control ABS'


corrupt_1 = pd.read_csv(corrupt_list[0], usecols=read_cols)
corrupt_1.columns = corrupt_1.columns.str.strip()
corrupt_1['ErrorBound'] = error_bounds[0]
corrupt_1['DecompressionBandwidth'] = (corrupt_1['DataSize']/corrupt_1['DecompressionTime'])/1048576
corrupt_1['RMSE'] = corrupt_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_2 = pd.read_csv(corrupt_list[1], usecols=read_cols)
corrupt_2.columns = corrupt_2.columns.str.strip()
corrupt_2['ErrorBound'] = error_bounds[1]
corrupt_2['DecompressionBandwidth'] = (corrupt_2['DataSize']/corrupt_2['DecompressionTime'])/1048576
corrupt_2['RMSE'] = corrupt_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_3 = pd.read_csv(corrupt_list[2], usecols=read_cols)
corrupt_3.columns = corrupt_3.columns.str.strip()
corrupt_3['ErrorBound'] = error_bounds[2]
corrupt_3['DecompressionBandwidth'] = (corrupt_3['DataSize']/corrupt_3['DecompressionTime'])/1048576
corrupt_3['RMSE'] = corrupt_3['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt = pd.concat([corrupt_1, corrupt_2, corrupt_3], ignore_index=True, sort=False)
corrupt['Type'] = 'Corrupt'
corrupt['Experiment'] = 'Corrupt ABS'

total = pd.concat([control, corrupt], ignore_index=True, sort=False)
abs_total = total.copy()
corrupt_completed = corrupt[(corrupt["Status"].str.contains('Completed')) | (corrupt["Status"].str.contains('Timeout'))]
```

### Control vs. Corrupt RMSE

```python
sns.barplot(x='ErrorBound', y='RMSE', hue="Type", data=corrupt_completed).set_title('RMSE by Error Bound')
plt.yscale('log')
plt.show()
```

### Control RMSE Breakdown

```python
print("ABS 1E2 Average RMSE: {}".format(control_1['RMSE'].mean()))
print("ABS 1E2 RMSE Standard Deviation: {}\n".format(control_1['RMSE'].std()))
print("ABS 1E3 Average RMSE: {}".format(control_2['RMSE'].mean()))
print("ABS 1E3 RMSE Standard Deviation: {}\n".format(control_2['RMSE'].std()))
print("ABS 1E4 Average RMSE: {}".format(control_3['RMSE'].mean()))
print("ABS 1E4 RMSE Standard Deviation: {}\n".format(control_3['RMSE'].std()))
```

### Corrupt Maximum Difference Breakdown

```python
corrupt_1_completed = corrupt_1[(corrupt_1["Status"].str.contains('Completed'))]
corrupt_2_completed = corrupt_2[(corrupt_2["Status"].str.contains('Completed'))]
corrupt_3_completed = corrupt_3[(corrupt_3["Status"].str.contains('Completed'))]

print("ABS 1E2 Average RMSE: {}".format(corrupt_1_completed['RMSE'].mean()))
print("ABS 1E2 RMSE Standard Deviation: {}\n".format(corrupt_1_completed['RMSE'].std()))
print("ABS 1E3 Average RMSE: {}".format(corrupt_2_completed['RMSE'].mean()))
print("ABS 1E3 RMSE Standard Deviation: {}\n".format(corrupt_2_completed['RMSE'].std()))
print("ABS 1E4 Average RMSE: {}".format(corrupt_3_completed['RMSE'].mean()))
print("ABS 1E4 RMSE Standard Deviation: {}\n".format(corrupt_3_completed['RMSE'].std()))
```

Looking at these averages, we notice that there is a minor jump in the number of incorrect values between the control and corrupt cases. The standard deviation is much higher though leading to a high variance in the number of incorrect values that can be seen. Let's use a boxplot to see if we can see the full distribution.

```python
sns.boxplot(x='ErrorBound', y='RMSE', data=corrupt_completed).set_title('RMSE by Error Bound')
plt.yscale('log')
plt.show()
```

As expected, there are a ton of outliers which explains the high standard deviation. Therefore, in most cases only a small percentage of values will be incorrect, but there could be a ton of them as well. We can look to see if an increase in incorrect percentages has any effect on the decompression bandwidth.

```python
f, axes = plt.subplots(1, 2, figsize=(20, 7))
sns.scatterplot(x='RMSE', y='DecompressionBandwidth', data=corrupt_1_completed, ax=axes[0]).set_title('ABS 1e2: RMSE by Decompression Bandwidth')
sns.scatterplot(x='RMSE', y='DecompressionBandwidth', data=corrupt_1_completed, ax=axes[1]).set_title('ABS 1e2: RMSE by Decompression Bandwidth')
plt.yscale('log')
plt.show()

f, axes = plt.subplots(1, 2, figsize=(20, 7))
sns.scatterplot(x='RMSE', y='DecompressionBandwidth', data=corrupt_2_completed, ax=axes[0]).set_title('ABS 1e3: RMSE by Decompression Bandwidth')
sns.scatterplot(x='RMSE', y='DecompressionBandwidth', data=corrupt_2_completed, ax=axes[1]).set_title('ABS 1e3: RMSE by Decompression Bandwidth')
plt.yscale('log')
plt.show()

f, axes = plt.subplots(1, 2, figsize=(20, 7))
sns.scatterplot(x='RMSE', y='DecompressionBandwidth', data=corrupt_3_completed, ax=axes[0]).set_title('ABS 1e4: RMSE by Decompression Bandwidth')
sns.scatterplot(x='RMSE', y='DecompressionBandwidth', data=corrupt_3_completed, ax=axes[1]).set_title('ABS 1e4: RMSE by Decompression Bandwidth')
plt.yscale('log')
plt.show()
```

Looking at these results, it seems that the decompression bandwidth effects are centered in two main locations.

Looking at the results above we can see that there are two major distributions of RMSE when compared to Incorrect Percentage. Looking closer we can see the patterns of such.

Let's see if we can pin point some bit flip locations that lead to the largest Incorrect Percentage

```python
sns.scatterplot(x='ByteLocation', y='RMSE', data=corrupt_1_completed).set_title('ABS 1e2: Byte Location by RMSE')
plt.xscale('log')
plt.show()
```

```python
sns.scatterplot(x='ByteLocation', y='RMSE', data=corrupt_2_completed).set_title('ABS 1e3: Byte Location by RMSE')
plt.xscale('log')
plt.show()
```

```python
sns.scatterplot(x='ByteLocation', y='RMSE', data=corrupt_3_completed).set_title('ABS 1e4: Byte Location by RMSE')
plt.xscale('log')
plt.show()
```

From the above information we can see the same trend of distinct groups. 

```python
abs_1e2_corrupt_upper_outliers = corrupt_1_completed[corrupt_1_completed['RMSE'] > corrupt_1_completed.RMSE.quantile(0.75)]
abs_1e2_corrupt_lower_outliers = corrupt_1_completed[corrupt_1_completed['RMSE'] < corrupt_1_completed.RMSE.quantile(0.25)]
abs_1e3_corrupt_upper_outliers = corrupt_2_completed[corrupt_2_completed['RMSE'] > corrupt_2_completed.RMSE.quantile(0.75)]
abs_1e3_corrupt_lower_outliers = corrupt_2_completed[corrupt_2_completed['RMSE'] < corrupt_2_completed.RMSE.quantile(0.25)]
abs_1e4_corrupt_upper_outliers = corrupt_3_completed[corrupt_3_completed['RMSE'] > corrupt_3_completed.RMSE.quantile(0.75)]
abs_1e4_corrupt_lower_outliers = corrupt_3_completed[corrupt_3_completed['RMSE'] < corrupt_3_completed.RMSE.quantile(0.25)]

f, axes = plt.subplots(figsize=(12, 7))
fig = sns.kdeplot(abs_1e2_corrupt_upper_outliers['ByteLocation'], shade=True, label='1e-2')
sns.kdeplot(abs_1e3_corrupt_upper_outliers['ByteLocation'], shade=True, label='1e-3')
sns.kdeplot(abs_1e4_corrupt_upper_outliers['ByteLocation'], shade=True, label='1e-4')
fig.set_xlabel('Byte Location', fontsize=14, weight='bold')
fig.set_ylabel('Probability', fontsize=14, weight='bold')
fig.set_title('ABS: RMSE > Q3', fontsize=20)
plt.legend(title='ErrorBound')
plt.setp(fig.get_legend().get_texts(), fontsize='13')
plt.setp(fig.get_legend().get_title(), fontsize='15')
plt.tight_layout()
plt.xscale('log')
plt.show()

f, axes = plt.subplots(figsize=(12, 7))
fig = sns.kdeplot(abs_1e2_corrupt_lower_outliers['ByteLocation'], shade=True, label='1e-2')
sns.kdeplot(abs_1e3_corrupt_lower_outliers['ByteLocation'], shade=True, label='1e-3')
sns.kdeplot(abs_1e4_corrupt_lower_outliers['ByteLocation'], shade=True, label='1e-4')
fig.set_xlabel('Byte Location', fontsize=14, weight='bold')
fig.set_ylabel('Probability', fontsize=14, weight='bold')
fig.set_title('ABS: RMSE < Q1', fontsize=20)
plt.legend(title='ErrorBound')
plt.setp(fig.get_legend().get_texts(), fontsize='13')
plt.setp(fig.get_legend().get_title(), fontsize='15')
plt.tight_layout()
plt.xscale('log')
plt.show()
```

### Read in data

```python
control_list = psnr_control_list
corrupt_list = psnr_corrupt_list
error_bounds = ['060', '090', '120']

control_1 = pd.read_csv(control_list[0], usecols=read_cols)
control_1.columns = control_1.columns.str.strip()
control_1['ErrorBound'] = error_bounds[0]
control_1['DecompressionBandwidth'] = (control_1['DataSize']/control_1['DecompressionTime'])/1048576
control_1['RMSE'] = control_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_2 = pd.read_csv(control_list[1], usecols=read_cols)
control_2.columns = control_2.columns.str.strip()
control_2['ErrorBound'] = error_bounds[1]
control_2['DecompressionBandwidth'] = (control_2['DataSize']/control_2['DecompressionTime'])/1048576
control_2['RMSE'] = control_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control_3 = pd.read_csv(control_list[2], usecols=read_cols)
control_3.columns = control_3.columns.str.strip()
control_3['ErrorBound'] = error_bounds[2]
control_3['DecompressionBandwidth'] = (control_3['DataSize']/control_3['DecompressionTime'])/1048576
control_3['RMSE'] = control_3['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
control = pd.concat([control_1, control_2, control_3], ignore_index=True, sort=False)
control['Type'] = 'Control'
control['Experiment'] = 'Control PSNR'

corrupt_1 = pd.read_csv(corrupt_list[0], usecols=read_cols)
corrupt_1.columns = corrupt_1.columns.str.strip()
corrupt_1['ErrorBound'] = error_bounds[0]
corrupt_1['DecompressionBandwidth'] = (corrupt_1['DataSize']/corrupt_1['DecompressionTime'])/1048576
corrupt_1['RMSE'] = corrupt_1['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_2 = pd.read_csv(corrupt_list[1], usecols=read_cols)
corrupt_2.columns = corrupt_2.columns.str.strip()
corrupt_2['ErrorBound'] = error_bounds[1]
corrupt_2['DecompressionBandwidth'] = (corrupt_2['DataSize']/corrupt_2['DecompressionTime'])/1048576
corrupt_2['RMSE'] = corrupt_2['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt_3 = pd.read_csv(corrupt_list[2], usecols=read_cols)
corrupt_3.columns = corrupt_3.columns.str.strip()
corrupt_3['ErrorBound'] = error_bounds[2]
corrupt_3['DecompressionBandwidth'] = (corrupt_3['DataSize']/corrupt_3['DecompressionTime'])/1048576
corrupt_3['RMSE'] = corrupt_3['RMSE'].apply(pd.to_numeric, errors='coerce').fillna(-2)
corrupt = pd.concat([corrupt_1, corrupt_2, corrupt_3], ignore_index=True, sort=False)
corrupt['Type'] = 'Corrupt'
corrupt['Experiment'] = 'Corrupt PSNR'

total = pd.concat([control, corrupt], ignore_index=True, sort=False)
psnr_total = total.copy()
corrupt_completed = corrupt[(corrupt["Status"].str.contains('Completed')) | (corrupt["Status"].str.contains('Timeout'))]
```
